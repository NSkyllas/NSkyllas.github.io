## My model with 4 components: Support, distance, food, crosswinds

```r
log_chla <- as.matrix(log10(chla))
```

```r
# My formula
myCost05 <- function(wind.direction, wind.speed, target, type = "active", chla = chla$chlor_a, alpha, beta, gamma, delta) {
  
  ###################  
  # 1) Preprocessing:
  ###################
    
  # The difference between the wind direction and the bid's PDM
  dif <- abs(wind.direction - target)
  
  # Convert wind support to cost: Low wind support -> high cost, high wind support -> low cost
  # Method: 1) Subtract the 99th percentile from wind support values: Low support rasters now have large negative values, high support rasters are 
  # close to 0. 2) Take their absolute values: Low support rasters have now high positive values (high cost), high-support rasters are still close to 0
  # (low cost)
  costx <- abs(wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180) - quantile(wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180), 0.99))
  
  # Chla is hevily sqewed: Many zeros and a few high values. I take its logarithm, and rasters with > -0.5 (high chl a concentration) all get a 
  # (minimum) value of -0.5. Later on, in the actual formula (next step) I add 0.5 and take its absolute (like I did with support). Low-chl rasters
  # have high positive values (cost) while productive rasters are close to 0.
  log_chla <- as.matrix(log10(chla[!is.na(dif)]))
  log_chla[log_chla > -0.5] <- -0.5
  log_chla[is.na(log_chla)] <- 0
  
  ################################  
  # 2) The 4 formulas (components)
  ################################
    
  # Each coefficient takes a value between 0 and 1. Their sum has to be always = 1.
  a = alpha     # Wind support
  b = beta      # Distance
  c = gamma     # Food
  d = delta     # Crosswind
    
  # The actual formulas: a) Support is already pre-processed. b) For distance, I take the median value of support. All rasters are eual, so if a bird 
  # wants to follow the shortest great circle distance, it will follow a straight path with a b coefficient = 1. c) I "flip" the log-transformed chl a as
  # explained in the previous step. d) Crosswind component. If a bird wants to follow a path with low crosswinds (usually close to the coast), then a 
  # large d coefficient is necessary.
  dif[!is.na(dif)] <-  a * costx + 
                       b * quantile(costx, 0.5) +
                       c * 12 * abs(log_chla + 0.5) +
                       d * abs(wind.speed[!is.na(dif)] * sin(dif[!is.na(dif)] * pi/180))
    
  # Land rasters have a cost of 1000000. Arctic terns should not fly over land (in reality they do fly shortly over the UK, as far as I know).  
  dif[is.na(dif)] <- 1000000
#  dif[is.na(dif)] <- 0

  dif  
}
```

```r
# List of coefficients
s = 0
coefflist = c()
for (a in seq(0, 1, 0.1))
{
    for (b in seq(0, 1, 0.1))
    {
        for (c in seq(0, 1, 0.1))
        {
            for (d in seq(0, 1, 0.1))
            {if (a + b + c + d == 1)
                {s = s + 1
                coefflist = c(coefflist, c(a, b, c, d))}
            }
        }
    }    
}
s
```

```r
# List of all conductance surfaces
mylist <- c()
for (a in seq(0, 1, 0.1))
{
    for (b in seq(0, 1, 0.1))
    {
        for (c in seq(0, 1, 0.1))
        {
            for (d in seq(0, 1, 0.1))
            {if (a + b + c + d == 1)
                {mylist <- c(mylist, flow.dispersion(my_raster, fun = myCost05, output = "transitionLayer", type = "active", chla = chla$chlor_a, a, b, c, d))}
            }
        }
    }    
}
```

```r
# Geocorrected conductance surfaces
mylistC <- c()
for (m in mylist)
    {mylistC <- c(mylistC, geoCorrection(m))
    }
```

```r
# List of shortest paths (SpatialLines) for Svalbard birds
A <- c(-25, -73) 
B <- c(-20, 65)
SSlist <- c()
for (m in mylist)
    {SSlist <- c(SSlist, shortestPath(m, A, B, output = "SpatialLines") )
    }
```

### Non-corrected

```r
# Extract coordinates from SpatialLines
SScoordlist <- c()
for (i in SSlist)
{
  SScoordlist <- c(SScoordlist, coordinates(i))
}
```

```r
# Extract coordinates from SpatialLines
NScoordlist <- c()
for (i in NSlist)
{
  NScoordlist <- c(NScoordlist, coordinates(i))
}
```

```r
# Take the mean (estimated by my model) lon per degree of lat 
SScoordmeanlist <- list()
for (i in seq(1, length(coefflist)/4, 1))
{
  SScoordmeanlist[[i]] <- as.data.frame(aggregate(x ~ y, geom(SSlist[[i]]), mean))
}
```

```r
# Take the mean (estimated by my model) lon per degree of lat 
NScoordmeanlist <- list()
for (i in seq(1, length(coefflist)/4, 1))
{
  NScoordmeanlist[[i]] <- as.data.frame(aggregate(x ~ y, geom(NSlist[[i]]), mean))
}
```

```r
# Add an extra column: 3-degree bins
SScoordmeanlist3 <- list()
for (i in seq(1, length(coefflist)/4, 1))
    {
        SScoordmeanlist[[i]]['lat_bins3'] <- apply(SScoordmeanlist[[i]]['y'], 2, cut, seq(-79, 80, 3)) 
    }
```

```r
# Add an extra column: 3-degree bins
NScoordmeanlist3 <- list()
for (i in seq(1, length(coefflist)/4, 1))
    {
        NScoordmeanlist[[i]]['lat_bins3'] <- apply(NScoordmeanlist[[i]]['y'], 2, cut, seq(-79, 80, 3)) 
    }
```

```r
# Average the calculated lon per 3-degree bin
SScoordbins3 <- list()
for (i in seq(1, length(coefflist)/4, 1))
{
  SScoordbins3[[i]] <- as.data.frame(aggregate(x ~ lat_bins3, SScoordmeanlist[[i]], mean))
}
```

```r
# Average the calculated lon per 3-degree bin
NScoordbins3 <- list()
for (i in seq(1, length(coefflist)/4, 1))
{
  NScoordbins3[[i]] <- as.data.frame(aggregate(x ~ lat_bins3, NScoordmeanlist[[i]], mean))
}
```

```r
# Change the bin labels of the observations (geolocators -> averaged per 3 degrees)
bird_SS_data$lat_bins <- (apply(bird_SS_data['lat_bins'], 2, FUN = function (x) gsub(" ", "", x, fixed = TRUE)))
```

```r
# Change the bin labels of the observations (geolocators -> averaged per 3 degrees)
bird_NS_data$lat_bins <- (apply(bird_NS_data['lat_bins'], 2, FUN = function (x) gsub(" ", "", x, fixed = TRUE)))
```

```r
# Rename columns to match the model output
colnames(bird_SS_data) <- c('lat_bins3', 'lon', 'lat', 'geometry')
```

```r
# Rename columns to match the model output
colnames(bird_NS_data) <- c('lat_bins3', 'lon', 'lat', 'geometry')
```

```r
# Merge observations and model by 3-degree bin column
SScoordmerged <- list()
for (i in seq(1, length(coefflist)/4, 1))
{
  SScoordmerged[[i]] <-  merge(bird_SS_data, SScoordbins3[[i]])
}
```

```r
# Merge observations and model by 3-degree bin column
NScoordmerged <- list()
for (i in seq(1, length(coefflist)/4, 1))
{
  NScoordmerged[[i]] <-  merge(bird_NS_data, NScoordbins3[[i]])
}
```

```r
# Keep only the necessary columns
for (i in seq(1, length(coefflist)/4, 1))
{
  SScoordmerged[[i]] <-  SScoordmerged[[i]][c('lat', 'lon', 'x')]
}
```

```r
# Keep only the necessary columns
for (i in seq(1, length(coefflist)/4, 1))
{
  NScoordmerged[[i]] <-  NScoordmerged[[i]][c('lat', 'lon', 'x')]
}
```

```r
# Order using lat
for (i in seq(1, length(coefflist)/4, 1))
{
  SScoordmerged[[i]] <-  SScoordmerged[[i]][order(SScoordmerged[[i]]['lat']),]
}
```

```r
# Order using lat
for (i in seq(1, length(coefflist)/4, 1))
{
  NScoordmerged[[i]] <-  NScoordmerged[[i]][order(NScoordmerged[[i]]['lat']),]
}
```

```r
#coord_merged <- coord_merged[order(coord_merged['lat']),]
```

```r
# Create 4 vectors with the 4 coefficients
alpha <- coefflist[seq(1, length(coefflist), 4)]
beta <- coefflist[seq(2, length(coefflist), 4)]
gamma <- coefflist[seq(3, length(coefflist), 4)]
delta <- coefflist[seq(4, length(coefflist), 4)]
```

```r
# Calculate statistics (RMSE, SD, corr. coeff) for each simulated shortest path
RMSE <- c()
R <- c()
SD_diff <- c()

for (i in seq(1, length(coefflist)/4, 1))
{
    RMSE <- c(RMSE, sqrt(mean((SScoordmerged[[i]]$x - SScoordmerged[[i]]$lon)**2)))
    R <- c(R, cor(SScoordmerged[[i]]$x, SScoordmerged[[i]]$lon))
    SD_diff <- c(SD_diff, abs(sd(SScoordmerged[[i]]$x) - sd(SScoordmerged[[i]]$lon)))
}
df_SS <- data.frame(a = alpha, b = beta, c = gamma, d = delta, rmse = RMSE, r = R, sd_diff = SD_diff)
```

```r
# Remove NAs for the Dutch dataset
for (i in seq(1, length(coefflist)/4, 1))
{
  NScoordmerged[[i]] <-  NScoordmerged[[i]][complete.cases(NScoordmerged[[i]]), ]
}
```

```r
# Calculate statistics (RMSE, SD, corr. coeff) for each simulated shortest path
RMSE <- c()
R <- c()
SD_diff <- c()

for (i in seq(1, length(coefflist)/4, 1))
{
    RMSE <- c(RMSE, sqrt(mean((NScoordmerged[[i]]$x - NScoordmerged[[i]]$lon)**2)))
    R <- c(R, cor(NScoordmerged[[i]]$x, NScoordmerged[[i]]$lon))
    SD_diff <- c(SD_diff, abs(sd(NScoordmerged[[i]]$x) - sd(NScoordmerged[[i]]$lon)))
}
df_NS <- data.frame(a = alpha, b = beta, c = gamma, d = delta, rmse = RMSE, r = R, sd_diff = SD_diff)
```
