```r
# My formula
# if dif > 90 : Cost = α*|support| + β*crosswind
# if dif <= 90: Cost = β*crosswind

myCost <- function(wind.direction, wind.speed, target, type = "active", alpha) {
  dif <- (abs(wind.direction - target))
  # If dif > 180 and is not NA
  dif[dif > 180 & !is.na(dif)] <- 360 - dif[dif > 180 & !is.na(dif)] # Modified from the original function
  wind.speed[is.na(wind.speed)] <- 0
  if (type == "passive") {
    # In "passive" type, if dif > 90, movement is not allowed
    dif[dif >= 90 & !is.na(dif)] <- Inf # check
    # For sea currents, dif could be NA is there are lands around
    dif[is.na(dif)] <- Inf # Modified from the original function
    # Here we apply the formula in Felicísimo et al. 2008
    dif[dif < 90] <- abs(wind.speed[dif < 90] * sin(dif[dif < 90]*pi/180)) 
    dif[dif == 0] <- 0.1
  }
  else {
    # For sea currents, dif could be NA if there are lands around
    dif[is.na(dif)] <- Inf
    # For "active" type movements against flow are allowed
    
    dif[!is.na(dif) & dif > 90] <- alpha * abs(wind.speed[dif > 90] * cos(dif[dif > 90]*pi/180)) + abs(wind.speed[dif > 90] * sin(dif[dif > 90]*pi/180)) 
    dif[!is.na(dif) & dif <= 90] <- abs(wind.speed[dif <= 90] * sin(dif[dif <= 90]*pi/180))   
  }

}
```

```r
# My formula
# Cost = α*|sup.anomaly| + β*crosswind

myCostRe <- function(wind.direction, wind.speed, target, type = "active", alpha) {
  dif <- (abs(wind.direction - target))
  # If dif > 180 and is not NA
  dif[dif > 180 & !is.na(dif)] <- 360 - dif[dif > 180 & !is.na(dif)] # Modified from the original function
  wind.speed[is.na(wind.speed)] <- 0
  if (type == "passive") {
    # In "passive" type, if dif > 90, movement is not allowed
    dif[dif >= 90 & !is.na(dif)] <- Inf # check
    # For sea currents, dif could be NA is there are lands around
    dif[is.na(dif)] <- Inf # Modified from the original function
    # Here we apply the formula in Felicísimo et al. 2008
    dif[dif < 90] <- abs(wind.speed[dif < 90] * sin(dif[dif < 90])) 
    dif[dif == 0] <- 0.1
  }
  else {
    # For sea currents, dif could be NA if there are lands around
    dif[is.na(dif)] <- Inf
    # For "active" type movements against flow are allowed
    dif[!is.na(dif)] <- alpha * abs((wind.speed * cos(dif * pi/180)) - max((wind.speed * cos(dif * pi/180)))) + 1 * abs(wind.speed * sin(dif * pi/180)) 
  }
}
```

```r
# My formula
# Cost = α*|sup.anomaly| + β*crosswind

myCostReb <- function(wind.direction, wind.speed, target, type = "active", alpha) {
  dif <- (abs(wind.direction - target))
  # If dif > 180 and is not NA
  dif[dif > 180 & !is.na(dif)] <- 360 - dif[dif > 180 & !is.na(dif)] # Modified from the original function
  wind.speed[is.na(wind.speed)] <- 0
  if (type == "passive") {
    # In "passive" type, if dif > 90, movement is not allowed
    dif[dif >= 90 & !is.na(dif)] <- Inf # check
    # For sea currents, dif could be NA is there are lands around
    dif[is.na(dif)] <- Inf # Modified from the original function
    # Here we apply the formula in Felicísimo et al. 2008
    dif[dif < 90] <- abs(wind.speed[dif < 90] * sin(dif[dif < 90])) 
    dif[dif == 0] <- 0.1
  }
  else {
    # For sea currents, dif could be NA if there are lands around
    dif[is.na(dif)] <- Inf
    # For "active" type movements against flow are allowed

    dif[!is.na(dif)] <- alpha * abs((wind.speed * cos(dif * pi/180)) - max(values(wind.speed * cos(dif * pi/180)))) + 1 * abs(wind.speed * sin(dif * pi/180)) 
    #dif[!is.na(dif)] <- 2 * dif[!is.na(dif)]
    #dif[dif == 0] <- 0.1
  }

#  dif * wind.speed
}
```

```r
# My formula
# if dif > 90 : Cost = α*|support| + β*crosswind
# if dif <= 90: Cost = β*crosswind

myCost <- function(wind.direction, wind.speed, target, type = "active") {
  dif <- (abs(wind.direction - target))
  # If dif > 180 and is not NA
  dif[dif > 180 & !is.na(dif)] <- 360 - dif[dif > 180 & !is.na(dif)] # Modified from the original function
  #dif[is.na(dif)] <- Inf
  #wind.speed[is.na(wind.speed)] <- 0
    
  if (type == "passive") {
    # In "passive" type, if dif > 90, movement is not allowed
    dif[dif >= 90 & !is.na(dif)] <- Inf # check
    # For sea currents, dif could be NA is there are lands around
    dif[is.na(dif)] <- Inf # Modified from the original function
    # Here we apply the formula in Felicísimo et al. 2008
    dif[dif < 90] <- abs(wind.speed[dif < 90] * sin(dif[dif < 90]*pi/180)) 
    dif[dif == 0] <- 0.1
  }
  
  else {
    # For sea currents, dif could be NA if there are lands around
    #dif[is.na(dif)] <- Inf
    # For "active" type movements against flow are allowed
    
    dif[!is.na(dif) & dif > 90] <- 10 * abs(wind.speed[!is.na(dif) & dif > 90] * cos(dif[!is.na(dif) & dif > 90]*pi/180)) + 
      abs(wind.speed[!is.na(dif) & dif > 90] * sin(dif[!is.na(dif) & dif > 90]*pi/180)) 
   
    dif[!is.na(dif) & dif <= 90] <- abs(wind.speed[!is.na(dif) & dif <= 90] * sin(dif[!is.na(dif) & dif <= 90]*pi/180))  
    dif[is.na(dif)] <- Inf
    #dif[is.infinite(dif)] <- Inf
  }

}
```

```r
# Cost computation following Muñoz et al., 2004; Felicísimo et al., 2008
#' @rdname flow.dispersion
#' @export
old_cost <- function(wind.direction, wind.speed, target, type = "Active") {
  dif <- (abs(wind.direction - target))
  # If dif > 180 and is not NA
  dif[dif > 180 & !is.na(dif)] <- 360 - dif[dif > 180 & !is.na(dif)] # Modified from the original function
  if (type == "passive") {
    # In "passive" type, if dif > 90, movement is not allowed
    dif[dif >= 90 & !is.na(dif)] <- Inf # check
    # For sea currents, dif could be NA is there are lands around
    dif[is.na(dif)] <- Inf # Modified from the original function
    # Here we apply the formula in Felicísimo et al. 2008
    dif[dif < 90] <- 2 * dif[dif < 90]
    dif[dif == 0] <- 0.1
  }
  else {
      #print("Only passive movements are currently allowed")
    # For sea currents, dif could be NA is there are lands around
    #dif[is.na(dif)] <- Inf
    # For "active" type movements against flow are allowed, so simply we
    # multiply by 2 the dif, following Felicísimo et al., 2008
    wind.speed[is.na(wind.speed)] <- 0  
    dif[!is.na(dif)] <- 2 * dif[!is.na(dif)] / wind.speed[!is.na(dif)]
    dif[dif == 0] <- 0.1
    dif[is.na(dif)] <- 1000000  
  }

  #wind.speed[is.na(wind.speed)] <- 0
  #dif / wind.speed
  dif
}
```

## gDistance example 1

```r
r <- raster(system.file("external/maungawhau.grd", package = "gdistance"))
```

```r
r
```

```r
plot(r)
```

```r
altDiff <- function(x) x[2] - x[1]
hd <- transition(r, altDiff, 8, symm= FALSE)
```

```r
altDiff
```

```r
hd
```

```r
image(transitionMatrix(hd))
```

```r
slope <- geoCorrection(hd)
```

```r
image(transitionMatrix(slope))
```

```r
adj <- adjacent(r, cells = 1:ncell(r), pairs = TRUE, directions = 8) 
speed <- slope 
speed[adj] <- 6 * exp(-3.5 * abs(slope[adj] + 0.05))
```

```r
Conductance <- geoCorrection(speed)
```

```r
image(transitionMatrix(Conductance))
```

```r
A <- c(2667670, 6479000) 
B <- c(2667800, 6479400)
AtoB <- shortestPath(Conductance, A, B, output = "SpatialLines") 
BtoA <- shortestPath(Conductance, B, A, output = "SpatialLines")
```

```r
plot(AtoB)
```

```r
plot(BtoA)
```

```r
plot(r, xlab = "x coordinate (m)", ylab = "y coordinate (m)", legend.lab = "Altitude (masl)") 
lines(AtoB, col = "red", lwd = 2) 
lines(BtoA, col = "blue")
text(A[1] - 10, A[2] - 10, "A") 
text(B[1] + 10, B[2] + 10, "B")
```
