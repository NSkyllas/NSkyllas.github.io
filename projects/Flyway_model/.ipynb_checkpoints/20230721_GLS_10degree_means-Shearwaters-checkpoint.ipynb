```python
dfc = pd.read_csv('GPS/Dataset_14907_518_Sooty_shearwater_geolocators_f_imported.csv')
```

```python
dfm['longitude'] = dfm.longitude.where(dfm.longitude > 0, dfm.longitude + 360)
dfc['longitude'] = dfc.longitude.where(dfc.longitude > 0, dfc.longitude + 360)
```

```python
gdfm = gpd.GeoDataFrame(dfm, geometry=gpd.points_from_xy(dfm.longitude, dfm.latitude), crs="EPSG:4326")
gdfc = gpd.GeoDataFrame(dfc, geometry=gpd.points_from_xy(dfc.longitude, dfc.latitude), crs="EPSG:4326")
```

```python
gdfm['date_gmt'] = pd.to_datetime(gdfm.date_gmt)
gdfc['date_gmt'] = pd.to_datetime(gdfc.date_gmt)
```

```python
gdfc['season'] = gdfc.date_gmt.dt.month < 7
gdfm['season'] = gdfm.date_gmt.dt.month < 7
```

```python
gdfc = gdfc.replace({'season': {True: 'spring', False: 'autumn'}})
gdfm = gdfm.replace({'season': {True: 'spring', False: 'autumn'}})
```

```python
gdfm = gdfm[gdfm.breed_stage == 'migration']
gdfc = gdfc[(gdfc.breed_stage == 'migration') & (~gdfc.bird_id.isin(range(1, 8)))]
```

```python
gdfc.date_gmt.dt.month.plot.hist(bins = 10)
gdfm.date_gmt.dt.month.plot.hist(bins = 10)
```

```python
gdfm = gdfm.sort_values(by = 'date_gmt')
gdfc = gdfc.sort_values(by = 'date_gmt')
```

```python
gdfm2 = gpd.GeoDataFrame(gdfm.groupby(['original_track_id', 'season'])['geometry'].apply(lambda x: LineString(x.tolist())))
gdfc2 = gpd.GeoDataFrame(gdfc.groupby(['original_track_id', 'season'])['geometry'].apply(lambda x: LineString(x.tolist())))
```

```python
gdfm2 = gdfm2.reset_index()
gdfc2 = gdfc2.reset_index()
```

```python
gdfm2 = gdfm2[gdfm2.original_track_id%2 == 1]
gdfc2 = gdfc2[gdfc2.original_track_id%2 == 1]
```

```python
gdf = gpd.GeoDataFrame(pd.concat([gdfc, gdfm]), crs="EPSG:4326")
```

```python
gdf = gdf.sort_values(by = 'latitude')
```

```python
gdf['lat_bin5'] = pd.cut(gdf.latitude, np.arange(-80, 80, 5))#, labels = np.arange(-80, 80, 5))
gdf['lat_bin10'] = pd.cut(gdf.latitude, np.arange(-80, 80, 10))#, labels = np.arange(-80, 80, 10))
```

```python
gdf_js = gdf[(gdf.wintering == 'Jap') & (gdf.season == 'spring')]
gdf_as = gdf[(gdf.wintering == 'Al') & (gdf.season == 'spring')]
gdf_cs = gdf[(gdf.wintering == 'Cal') & (gdf.season == 'spring')]
gdf_ja = gdf[(gdf.wintering == 'Jap') & (gdf.season == 'autumn')]
gdf_aa = gdf[(gdf.wintering == 'Al') & (gdf.season == 'autumn')]
gdf_ca = gdf[(gdf.wintering == 'Cal') & (gdf.season == 'autumn')]
```

```python
gdf_ja.plot()
```

```python
for comb in ['js', 'ja', 'as', 'aa', 'cs', 'ca']:
             exec("gdf_{0}_10 = pd.DataFrame(data = np.array([gdf_{0}.groupby('lat_bin10')[['latitude']].median().values.ravel(), gdf_{0}.groupby('lat_bin10')[['longitude']].median().values.ravel(), gdf_{0}.groupby('lat_bin10')[['longitude']].std().values.ravel(), gdf_{0}.groupby('lat_bin10')[['longitude']].std().index]))".format(comb))
```

```python
for comb in ['js', 'ja', 'as', 'aa', 'cs', 'ca']:
    exec("gdf_{0}_10 = gdf_{0}_10.T".format(comb))
    exec("gdf_{0}_10.columns = ['lat_median10', 'lon_median10', 'lon_sd_10', 'lat_bins_10']".format(comb))
```

```python
gdf_js_10.iloc[0, 1] = dfc.lon_colony.unique()[0]
gdf_as_10.iloc[0, 1] = dfc.lon_colony.unique()[0]
gdf_cs_10.iloc[0, 1] = dfc.lon_colony.unique()[0]
```

```python
gdf_js_10.iloc[0, 0] = dfc.lat_colony.unique()[0]
gdf_as_10.iloc[0, 0] = dfc.lat_colony.unique()[0]
gdf_cs_10.iloc[1, 0] = dfc.lat_colony.unique()[0]
```

```python
spring = gdf_js_10
autumn = gdf_ja_10

fig, ax = plt.subplots()
gdf_js.plot(ax = ax, color = 'tab:blue', alpha = 0.4)
gdf_ja.plot(ax = ax, color = 'tab:orange', alpha = 0.4)
ax.plot(spring.lon_median10, spring.lat_median10)
ax.plot(autumn.lon_median10, autumn.lat_median10)
ax.fill_betweenx(spring.lat_median10.values.astype(float), 
                  (spring.lon_median10 - spring.lon_sd_10).values.astype(float), 
                  (spring.lon_median10 + spring.lon_sd_10).values.astype(float), 
                  color = 'tab:blue', alpha = 0.5)
ax.fill_betweenx(autumn.lat_median10.values.astype(float), 
                  (autumn.lon_median10 - autumn.lon_sd_10).values.astype(float), 
                  (autumn.lon_median10 + autumn.lon_sd_10).values.astype(float), 
                  color = 'tab:orange', alpha = 0.5)
ax.scatter(dfc.lon_colony, dfc.lat_colony, color = 'red', label = 'Codfish island', zorder = 3)
ax.scatter(dfm.lon_colony, dfm.lat_colony, color = 'green', label = 'Mana island', zorder = 3)
ax.legend()
```
