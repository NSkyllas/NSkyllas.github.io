```python
# Load the list with all 159 coefficient combinations and split it to 4 columns
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("alist_{0} = [coef_{0}.iloc[i, :].values[1] for i in range(0, len(coef_{0}), 4)]".format(comb1, comb1))
    exec("blist_{0} = [coef_{0}.iloc[i+1, :].values[1] for i in range(0, len(coef_{0}), 4)]".format(comb1, comb1))
    exec("clist_{0} = [coef_{0}.iloc[i+2, :].values[1] for i in range(0, len(coef_{0}), 4)]".format(comb1, comb1))
    exec("dlist_{0} = [coef_{0}.iloc[i+3, :].values[1] for i in range(0, len(coef_{0}), 4)]".format(comb1, comb1))
# Turn the coefficient list into a DataFrame with 4 columns
    exec("df_coef_{0} = pd.DataFrame([alist_{0}, blist_{0}, clist_{0}, dlist_{0}])".format(comb1, comb2))
    exec("df_coef_{0} = df_coef_{0}.T".format(comb1))
    exec("df_coef_{0}.columns = ['a', 'b', 'c', 'd']".format(comb1))
    exec("df_coef_{0}['matrix'] = df_coef_{0}.index + 1".format(comb1))
```

```python
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("dfsaam_{0} = dfsaam_{0}.merge(df_coef_{0}, on = 'matrix')".format(comb1))
    exec("dfsaaf_{0} = dfsaaf_{0}.merge(df_coef_{0}, on = 'matrix')".format(comb1))
    exec("dfna_{0} = dfna_{0}.merge(df_coef_{0}, on = 'matrix')".format(comb1))
```

```python
# Turn it to a GeoDataFrame
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam_{0} = gpd.GeoDataFrame(dfsaam_{0}, geometry = gpd.points_from_xy(dfsaam_{0}['x'], dfsaam_{0}['y']))".format(comb1))
    exec("gdfsaaf_{0} = gpd.GeoDataFrame(dfsaaf_{0}, geometry = gpd.points_from_xy(dfsaaf_{0}['x'], dfsaaf_{0}['y']))".format(comb1))
    exec("gdfna_{0} = gpd.GeoDataFrame(dfna_{0}, geometry = gpd.points_from_xy(dfna_{0}['x'], dfna_{0}['y']))".format(comb1))
```

```python
# Collect all the coordinates and put them in 1 list per path
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam2_{0} = gpd.GeoDataFrame(gdfsaam_{0}.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))".format(comb1))
    exec("gdfsaaf2_{0} = gpd.GeoDataFrame(gdfsaaf_{0}.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))".format(comb1))
    exec("gdfna2_{0} = gpd.GeoDataFrame(gdfna_{0}.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))".format(comb1))
```

```python
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    # Add the coefficient combinations from the other DataFrame
    exec("gdfsaam2_{0} = gdfsaam2_{0}.merge(gdfsaam_{0}, left_on = gdfsaam2_{0}.index, right_on = 'matrix')".format(comb1))
    exec("gdfsaaf2_{0} = gdfsaaf2_{0}.merge(gdfsaaf_{0}, left_on = gdfsaaf2_{0}.index, right_on = 'matrix')".format(comb1))
    exec("gdfna2_{0} = gdfna2_{0}.merge(gdfna_{0}, left_on = gdfna2_{0}.index, right_on = 'matrix')".format(comb1))
```

```python
# Drop duplicate columns
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam2_{0} = gdfsaam2_{0}.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = dict.fromkeys(['geometry_x'], 'geometry'))".format(comb1))
    exec("gdfsaaf2_{0} = gdfsaaf2_{0}.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = dict.fromkeys(['geometry_x'], 'geometry'))".format(comb1))
    exec("gdfna2_{0} = gdfna2_{0}.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = dict.fromkeys(['geometry_x'], 'geometry'))".format(comb1))
```

```python
# Turn back to a dataframe
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam2_{0} = gpd.GeoDataFrame(gdfsaam2_{0}, crs=4326)".format(comb1))
    exec("gdfsaaf2_{0} = gpd.GeoDataFrame(gdfsaaf2_{0}, crs=4326)".format(comb1))
    exec("gdfna2_{0} = gpd.GeoDataFrame(gdfna2_{0}, crs=4326)".format(comb1))
```

```python
# Calculate each path's length (in km)
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam2_{0}['length_km'] = gdfsaam2_{0}.geometry.to_crs(3857).length/1000".format(comb1))
    exec("gdfsaaf2_{0}['length_km'] = gdfsaaf2_{0}.geometry.to_crs(3857).length/1000".format(comb1))
    exec("gdfna2_{0}['length_km'] = gdfna2_{0}.geometry.to_crs(3857).length/1000".format(comb1))
```

```python
# Create an extra column with bins; one bin per 10 degrees of latitude
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam_{0}['lat_bins'] = pd.cut(gdfsaam_{0}.y, bins = range(-80, 80, 10))".format(comb1))
    exec("gdfsaaf_{0}['lat_bins'] = pd.cut(gdfsaaf_{0}.y, bins = range(-80, 80, 10))".format(comb1))
    exec("gdfna_{0}['lat_bins'] = pd.cut(gdfna_{0}.y, bins = range(-80, 80, 10))".format(comb1))
```

```python
# Take the mean position of the bird for each 3-degree bin
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("gdfsaam3_{0} = gdfsaam_{0}.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).median().reset_index()".format(comb1))
    exec("gdfsaaf3_{0} = gdfsaaf_{0}.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).median().reset_index()".format(comb1))
    exec("gdfna3_{0} = gdfna_{0}.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).median().reset_index()".format(comb1))
```

```python
# Remove the bins with NaNs (the ones without any datapoints, cannot generate a distance or RMSE)
# Perhaps replace them with zeros instead of removing them?
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
     exec("gdfsaam3_{0} = gdfsaam3_{0}[gdfsaam3_{0}.lat_bins.astype(str) != '(-80, -70]']".format(comb1))
     exec("gdfsaaf3_{0} = gdfsaaf3_{0}[gdfsaaf3_{0}.lat_bins.astype(str) != '(-80, -70]']".format(comb1))
     exec("gdfna3_{0} = gdfna3_{0}[(gdfna3_{0}.lat_bins.astype(str) != '(-80, -70]') & (gdfna3_{0}.lat_bins.astype(str) != '(-70, -60]') & (gdfna3_{0}.lat_bins.astype(str) != '(-60, -50]') & (gdfna3_{0}.lat_bins.astype(str) != '(60, 70]')]".format(comb1))
bird_NA = bird_NA[bird_NA.lat_bins_10.astype(str) != '(-70, -60]']
```

```python
# For each matrix (path), calcuate the distance between its points and the reference (bird tracking data). 
# for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
def dist_hav_sa(data, data2):
    dist_bird = np.array([])
    for m in data.matrix.unique():
        for i in range(14):
            dist_bird = np.append(dist_bird, haversine((data2.lat_median10.values[i], data[data.matrix == m].x.values[i]), (data2.lat_median10.values[i], data2.lon_median10.values[i])))
    data['error'] = dist_bird
```

```python
# For each matrix (path), calcuate the distance between its points and the reference (bird tracking data). 
# for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
def dist_hav_na(data, data2):
    dist_bird = np.array([])
    for m in data.matrix.unique():
        for i in range(11):
            dist_bird = np.append(dist_bird, haversine((data2.lat_median10.values[i], data[data.matrix == m].x.values[i]), (data2.lat_median10.values[i], data2.lon_median10.values[i])))
    data['error'] = dist_bird
```

```python
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("dist_hav_sa(gdfsaam3_{0}, bird_SA_am)".format(comb1))
    exec("dist_hav_sa(gdfsaaf3_{0}, bird_SA_af)".format(comb1))
    exec("dist_hav_na(gdfna3_{0}, bird_NA)".format(comb1))
```

```python
# For each path (matrix), calculate the RMSE (square root of the mean of the squared errors)
# RMSE will be our main metric for finding the best path
def rmse_hav(data, data2):
    rmse = np.array([])
    for m in data.matrix.unique():
        rmse = np.append(rmse, np.sqrt(np.mean((data[data.matrix == m].error**2))))
    data2['rmse'] = rmse
```

```python
for comb1 in ['00', '03', '05', '01', '30', '50', '10', '33', '55', '11']:
    exec("rmse_hav(gdfsaam3_{0}, gdfsaam2_{0})".format(comb1))
    exec("rmse_hav(gdfsaaf3_{0}, gdfsaaf2_{0})".format(comb1))
    exec("rmse_hav(gdfna3_{0}, gdfna2_{0})".format(comb1))
```
