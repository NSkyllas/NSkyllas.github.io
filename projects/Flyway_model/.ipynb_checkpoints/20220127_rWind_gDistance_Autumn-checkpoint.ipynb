```r
# Cost computation following Muñoz et al., 2004; Felicísimo et al., 2008
#' @rdname flow.dispersion
#' @export
old_cost <- function(wind.direction, wind.speed, target, type = "Active") {
  dif <- (abs(wind.direction - target))
  # If dif > 180 and is not NA
  dif[dif > 180 & !is.na(dif)] <- 360 - dif[dif > 180 & !is.na(dif)] # Modified from the original function
  if (type == "passive") {
    # In "passive" type, if dif > 90, movement is not allowed
    dif[dif >= 90 & !is.na(dif)] <- Inf # check
    # For sea currents, dif could be NA is there are lands around
    dif[is.na(dif)] <- Inf # Modified from the original function
    # Here we apply the formula in Felicísimo et al. 2008
    dif[dif < 90] <- 2 * dif[dif < 90]
    dif[dif == 0] <- 0.1
  }
  else {
      #print("Only passive movements are currently allowed")
    # For sea currents, dif could be NA is there are lands around
    #dif[is.na(dif)] <- Inf
    # For "active" type movements against flow are allowed, so simply we
    # multiply by 2 the dif, following Felicísimo et al., 2008
    wind.speed[is.na(wind.speed)] <- 0  
    dif[!is.na(dif)] <- 2 * dif[!is.na(dif)] / wind.speed[!is.na(dif)]
    dif[dif == 0] <- 0.1
    dif[is.na(dif)] <- 1000000  
  }

  #wind.speed[is.na(wind.speed)] <- 0
  #dif / wind.speed
  dif
}
```

```r
# My formula1
# if dif > 90 : Cost = α*|support| + β*crosswind
# if dif <= 90: Cost = β*crosswind

myCost01 <- function(wind.direction, wind.speed, target, type = "active") {
  dif <- abs(wind.direction - target)

  dif[!is.na(dif) & dif > 90] <- 0.9 * abs(wind.speed[!is.na(dif) & dif > 90] * cos(dif[!is.na(dif) & dif > 90]*pi/180)) + 
                                 0.1 * abs(wind.speed[!is.na(dif) & dif > 90] * sin(dif[!is.na(dif) & dif > 90]*pi/180)) 
  
  dif[!is.na(dif) & dif <= 90] <- 0.1 * abs(wind.speed[!is.na(dif) & dif <= 90] * sin(dif[!is.na(dif) & dif <= 90]*pi/180))  
    
  dif[is.na(dif)] <- 1000000

  dif  
}
```

```r
# Piersma10
# Cost = dfr.w * np.cos(np.radians(dfr.theta)) + np.sqrt(2 ** 2 + np.abs(dfr.w * np.sin(np.radians(dfr.theta))) ** 2) - 2
# Cost = np.abs(dfr.piersma2.where(dfr.piersma2 < dfr.piersma2.quantile(0.95), dfr.piersma2.quantile(0.95)) - dfr.piersma2.quantile(0.95))

piersma10 <- function(wind.direction, wind.speed, target, type = "active") {
  dif <- abs(wind.direction - target)

  dif[!is.na(dif)] <- wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180) +  sqrt(10 ** 2 + (abs(wind.speed[!is.na(dif)] * sin(dif[!is.na(dif)] * pi/180))) ** 2) - 10
  dif[!is.na(dif)] <- abs(dif[!is.na(dif)] - quantile(dif[!is.na(dif)], 0.95))  

    
  dif[is.na(dif)] <- 1000000

  dif  
}
```
