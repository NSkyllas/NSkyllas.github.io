## My model with 4 components: Support, distance, food, crosswinds

```r
log_chla <- as.matrix(log10(chla))
```

```r
# My formula
myCost05 <- function(wind.direction, wind.speed, target, type = "active", chla = chla$chlor_a, alpha, beta, gamma, delta) {
  
  ###################  
  # 1) Preprocessing:
  ###################
    
  # The difference between the wind direction and the bid's PDM
  dif <- abs(wind.direction - target)
  
  # Convert wind support to cost: Low wind support -> high cost, high wind support -> low cost
  # Method: 1) Subtract the 99th percentile from wind support values: Low support rasters now have large negative values, high support rasters are 
  # close to 0. 2) Take their absolute values: Low support rasters have now high positive values (high cost), high-support rasters are still close to 0
  # (low cost)
  costx <- abs(wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180) - quantile(wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180), 0.99))
  
  # Chla is hevily sqewed: Many zeros and a few high values. I take its logarithm, and rasters with > -0.5 (high chl a concentration) all get a 
  # (minimum) value of -0.5. Later on, in the actual formula (next step) I add 0.5 and take its absolute (like I did with support). Low-chl rasters
  # have high positive values (cost) while productive rasters are close to 0.
  log_chla <- as.matrix(log10(chla[!is.na(dif)]))
  log_chla[log_chla > -0.5] <- -0.5
  log_chla[is.na(log_chla)] <- 0
  
  ################################  
  # 2) The 4 formulas (components)
  ################################
    
  # Each coefficient takes a value between 0 and 1. Their sum has to be always = 1.
  a = alpha     # Wind support
  b = beta      # Distance
  c = gamma     # Food
  d = delta     # Crosswind
    
  # The actual formulas: a) Support is already pre-processed. b) For distance, I take the median value of support. All rasters are eual, so if a bird 
  # wants to follow the shortest great circle distance, it will follow a straight path with a b coefficient = 1. c) I "flip" the log-transformed chl a as
  # explained in the previous step. d) Crosswind component. If a bird wants to follow a path with low crosswinds (usually close to the coast), then a 
  # large d coefficient is necessary.
  dif[!is.na(dif)] <-  a * costx + 
                       b * quantile(costx, 0.5) +
                       c * 12 * abs(log_chla + 0.5) +
                       d * abs(wind.speed[!is.na(dif)] * sin(dif[!is.na(dif)] * pi/180))
    
  # Land rasters have a cost of 1000000. Arctic terns should not fly over land (in reality they do fly shortly over the UK, as far as I know).  
  dif[is.na(dif)] <- 1000000
#  dif[is.na(dif)] <- 0

  dif  
}
```

```r
# List of coefficients
s = 0
coefflist = c()
for (a in seq(0, 1, 0.1))
{
    for (b in seq(0, 0, 0.1))
    {
        for (c in seq(0, 0.5, 0.1))
        {
            for (d in seq(0, 0.5, 0.1))
            {if (a + b + c + d == 1)
                {s = s + 1
                coefflist = c(coefflist, c(a, b, c, d))}
            }
        }
    }    
}
s
```

```r
# List of all conductance surfaces
mylist <- c()
for (a in seq(0, 1, 0.1))
{
    for (b in seq(0, 0, 0.1))
    {
        for (c in seq(0, 0.5, 0.1))
        {
            for (d in seq(0, 0.5, 0.1))
            {if (a + b + c + d == 1)
                {mylist <- c(mylist, flow.dispersion(my_raster, fun = myCost05, output = "transitionLayer", type = "active", chla = chla$chlor_a, a, b, c, d))}
            }
        }
    }    
}
```

```r
# Geocorrected conductance surfaces
mylistC <- c()
for (m in mylist)
    {mylistC <- c(mylistC, geoCorrection(m))
    }
```

```r
# List of shortest paths (SpatialLines) for Svalbard AMAM birds
A <- c(10, 77)
B <- c(-20, -65)
SAlist_amam <- c()
for (m in mylistC)
    {SAlist_amam <- c(SAlist_amam, shortestPath(m, A, B, output = "SpatialLines") )
    }
```

```r
# Calculate statistics (RMSE, SD, corr. coeff) for each simulated shortest path
RMSE <- c()
R <- c()
SD_diff <- c()

for (i in seq(1, length(coefflist)/4, 1))
{
    distance <- mutate(SAcoordmerged_amam[[i]], distance = distHaversine(cbind(lon, lat), cbind(x, lat)))
    RMSE <- c(RMSE, sqrt(mean((distance$distance)**2)))
    R <- c(R, cor(SAcoordmerged_amam[[i]]$x, SAcoordmerged_amam[[i]]$lon))
    SD_diff <- c(SD_diff, abs(sd(SAcoordmerged_amam[[i]]$x) - sd(SAcoordmerged_amam[[i]]$lon)))
}
df_SA_amam <- data.frame(a = alpha, b = beta, c = gamma, d = delta, rmse = RMSE, r = R, sd_diff = SD_diff)
```

```r
# Calculate statistics (RMSE, SD, corr. coeff) for each simulated shortest path
RMSE <- c()
R <- c()
SD_diff <- c()

for (i in seq(1, length(coefflist)/4, 1))
{
    distance <- mutate(SAcoordmerged_afaf[[i]], distance = distHaversine(cbind(lon, lat), cbind(x, lat)))
    RMSE <- c(RMSE, sqrt(mean((distance$distance)**2)))
    R <- c(R, cor(SAcoordmerged_afaf[[i]]$x, SAcoordmerged_afaf[[i]]$lon))
    SD_diff <- c(SD_diff, abs(sd(SAcoordmerged_afaf[[i]]$x) - sd(SAcoordmerged_afaf[[i]]$lon)))
}
df_SA_afaf <- data.frame(a = alpha, b = beta, c = gamma, d = delta, rmse = RMSE, r = R, sd_diff = SD_diff)
```
