```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from cartopy import crs as ccrs
from shapely.geometry import Point, LineString, shape
import matplotlib.colors
import cartopy.feature as cfeature
from haversine import haversine
import numpy as np
```

```python
dfss = pd.read_csv('SScoordinates.csv')
dfns = pd.read_csv('NScoordinates.csv')
dfsa_am = pd.read_csv('SAamcoordinates.csv')
dfsa_af = pd.read_csv('SAafcoordinates.csv')
dfna = pd.read_csv('NAcoordinates.csv')
coef = pd.read_csv('coefflist.csv', index_col = 0)
```

```python
dfsa_af1 = pd.read_csv('SAafcoordinates1.csv')
dfsa_af2 = pd.read_csv('SAafcoordinates2.csv')
```

```python
# Load the list with all 159 coefficient combinations and split it to 4 columns
alist = [coef.iloc[i, :].values[0] for i in range(0, len(coef), 4)]
blist = [coef.iloc[i+1, :].values[0] for i in range(0, len(coef), 4)]
clist = [coef.iloc[i+2, :].values[0] for i in range(0, len(coef), 4)]
dlist = [coef.iloc[i+3, :].values[0] for i in range(0, len(coef), 4)]
```

```python
# Turn the coefficient list into a DataFrame with 4 columns
df_coef = pd.DataFrame({"a" : alist, 
                       "b": dlist, 
                       "c": blist, 
                       "d": clist})
```

```python
# Add an extra column: matrix (i.e. simulation, i.e. path number)
df_coef['matrix'] = df_coef.index + 1
```

```python
# Merge the coefficients with th coordinates of the 159 paths.
# The result is a DataFrame with all 159 paths: Coordinates per location, with coefficients & path number
dfss = dfss.merge(df_coef, on = 'matrix')
dfns = dfns.merge(df_coef, on = 'matrix')
dfsa_am = dfsa_am.merge(df_coef, on = 'matrix')
dfsa_af = dfsa_af.merge(df_coef, on = 'matrix')
dfna = dfna.merge(df_coef, on = 'matrix')
```

```python
dfsa_af1 = dfsa_af1.merge(df_coef, on = 'matrix')
dfsa_af2 = dfsa_af2.merge(df_coef, on = 'matrix')
```

```python
# Turn it to a GeoDataFrame
gdfss = gpd.GeoDataFrame(dfss, geometry = gpd.points_from_xy(dfss['x'], dfss['y']))
gdfsa_am = gpd.GeoDataFrame(dfsa_am, geometry = gpd.points_from_xy(dfsa_am['x'], dfsa_am['y']))
gdfsa_af = gpd.GeoDataFrame(dfsa_af, geometry = gpd.points_from_xy(dfsa_af['x'], dfsa_af['y']))
gdfns = gpd.GeoDataFrame(dfns, geometry = gpd.points_from_xy(dfns['x'], dfns['y']))
gdfna = gpd.GeoDataFrame(dfna, geometry = gpd.points_from_xy(dfna['x'], dfna['y']))
```

```python
# Collect all the coordinates and put them in 1 list per path
gdfss2 = gpd.GeoDataFrame(gdfss.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))
gdfns2 = gpd.GeoDataFrame(gdfns.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))
gdfsa_am2 = gpd.GeoDataFrame(gdfsa_am.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))
gdfsa_af2 = gpd.GeoDataFrame(gdfsa_af.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))
gdfna2 = gpd.GeoDataFrame(gdfna.groupby(['matrix'])['geometry'].apply(lambda x: LineString(x.tolist())))
```

```python
# Add the coefficient combinations from the other DataFrame
gdfss2 = gdfss2.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = {"geometry_x": "geometry"})
gdfns2 = gdfns2.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = {"geometry_x": "geometry"})
gdfsa_am2 = gdfsa_am2.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = {"geometry_x": "geometry"})
gdfsa_af2 = gdfsa_af2.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = {"geometry_x": "geometry"})
gdfna2 = gdfna2.drop_duplicates(subset = ['matrix']).drop(['x', 'y', 'geometry_y'], axis = 1).reset_index().drop('index', axis = 1).rename(columns = {"geometry_x": "geometry"})
```

```python
# Calculate each path's length (in km)
gdfss2['length_km'] = gdfss2.geometry.to_crs(3857).length/1000
gdfns2['length_km'] = gdfns2.geometry.to_crs(3857).length/1000
gdfsa_am2['length_km'] = gdfsa_am2.geometry.to_crs(3857).length/1000
gdfsa_af2['length_km'] = gdfsa_af2.geometry.to_crs(3857).length/1000
gdfna2['length_km'] = gdfna2.geometry.to_crs(3857).length/1000
```

```python
def hist_plot(data, pop, season):
    data.length_km.plot.hist(histtype = 'step', label = 'total')
    data[data.d >= 0.3].length_km.plot.hist(alpha = 0.4, label = 'd')
    data[data.b >= 0.3].length_km.plot.hist(alpha = 0.4, label = 'b')
    data[data.a >= 0.6].length_km.plot.hist(alpha = 0.7, label = 'a')
    data[data.c >= 0.6].length_km.plot.hist(alpha = 0.7, label = 'c')
    plt.xlim(14500, 34500)
    plt.legend()
    plt.title(str(pop) + " " + str(season))
    plt.show()
```

```python
def line_plot(data, pop, season):
    fig, ax = plt.subplots(ncols = 5, figsize = (22, 7))#, subplot_kw={'projection': ccrs.PlateCarree()})
    data[data['length_km'] < 50000].plot(ax = ax[0], column = 'a', alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data[data['length_km'] < 50000].plot(ax = ax[1], column = 'b', alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data[data['length_km'] < 50000].plot(ax = ax[2], column = 'c', alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data[data['length_km'] < 50000].plot(ax = ax[3], column = 'd', alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data[data['length_km'] < 50000].plot(ax = ax[4], column = 'length_km', alpha = 0.3, legend = True, linewidth = 2, cmap = 'PiYG_r')
    #ax[0].coastlines()
    #ax[1].coastlines()
    #ax[2].coastlines()
    #ax[3].coastlines()
    #ax[4].coastlines()
    #ax[0].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[1].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[2].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[3].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[4].add_feature(cfeature.OCEAN, facecolor = 'grey')
    ax[0].set_title('a')
    ax[1].set_title('b')
    ax[2].set_title('c')
    ax[3].set_title('d')
    ax[4].set_title('length_km')
    plt.suptitle(str(pop) + " " + str(season))
    plt.show()
```

```python
# Plot the length of each path. On top of the total histogram, coloured histograms depict paths in which alpha is high (red colour, b is high (purple) etc.
# Longer paths tend to have high alpha coefficients (wind support maximisation), shorter paths have high beta values (distance minimisation)
hist_plot(gdfss2[gdfss2.length_km < 50000], 'Svalbard', 'spring')
```

```python
# Let's load our tracking data, we will use them as reference for validating the model
bird_SS = pd.read_csv('../Least_cost_modelling/gdf_SS_10.csv')
bird_NS = pd.read_csv('../Least_cost_modelling/gdf_NS_10.csv') 
bird_SA_am = pd.read_csv('../Least_cost_modelling/gdf_SA_amam_10.csv')  
bird_SA_af = pd.read_csv('../Least_cost_modelling/gdf_SA_afaf_10.csv')  
bird_NA = pd.read_csv('../Least_cost_modelling/gdf_NA_10.csv')  
```

```python
# Turn them into a GeoDataFrame
bird_SS = gpd.GeoDataFrame(bird_SS, geometry = gpd.points_from_xy(bird_SS['lon_median10'], bird_SS['lat_median10']), crs=4326)
bird_NS = gpd.GeoDataFrame(bird_NS, geometry = gpd.points_from_xy(bird_NS['lon_median10'], bird_NS['lat_median10']), crs=4326)
bird_SA_am = gpd.GeoDataFrame(bird_SA_am, geometry = gpd.points_from_xy(bird_SA_am['lon_median10'], bird_SA_am['lat_median10']), crs=4326)
bird_SA_af = gpd.GeoDataFrame(bird_SA_af, geometry = gpd.points_from_xy(bird_SA_af['lon_median10'], bird_SA_af['lat_median10']), crs=4326)
bird_NA = gpd.GeoDataFrame(bird_NA, geometry = gpd.points_from_xy(bird_NA['lon_median10'], bird_NA['lat_median10']), crs=4326)
```

```python
bird_SS.sort_values(by = 'lat_median10')
```

```python
# Create an extra column with bins; one bin per 3 degrees of latitude
gdfss['lat_bins'] = pd.cut(gdfss.y, bins = range(-80, 80, 10))
gdfns['lat_bins'] = pd.cut(gdfns.y, bins = range(-80, 80, 10))
gdfsa_am['lat_bins'] = pd.cut(gdfsa_am.y, bins = range(-80, 80, 10))
gdfsa_af['lat_bins'] = pd.cut(gdfsa_af.y, bins = range(-80, 80, 10))
gdfna['lat_bins'] = pd.cut(gdfna.y, bins = range(-80, 80, 10))
```

```python
# Take the mean position of the bird for each 3-degree bin
gdfss3 = gdfss.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).median().reset_index()
gdfns3 = gdfns.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).median().reset_index()
gdfsa_am3 = gdfsa_am.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).mean().reset_index()
gdfsa_af3 = gdfsa_af.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).mean().reset_index()
gdfna3 = gdfna.drop('geometry', axis = 1).groupby(['matrix', 'lat_bins']).mean().reset_index()
```

```python
for i in range(195):
    plt.plot(gdfns3[gdfns3.matrix == i+1].x, 
             gdfns3[gdfns3.matrix == i+1].y, color = 'b', alpha = 0.1)
plt.plot(bird_NS.lon_median10, 
         bird_NS.lat_median10, linewidth = 3, linestyle = '--', zorder = 3, color = 'red')
plt.show()
```

```python
# Turn back to a geoDataFrame
gdfss3 = gpd.GeoDataFrame(gdfss3, geometry = gpd.points_from_xy(gdfss3['x'], gdfss3['y']), crs=4326)
gdfns3 = gpd.GeoDataFrame(gdfns3, geometry = gpd.points_from_xy(gdfns3['x'], gdfns3['y']), crs=4326)
gdfsa_am3 = gpd.GeoDataFrame(gdfsa_am3, geometry = gpd.points_from_xy(gdfsa_am3['x'], gdfsa_am3['y']), crs=4326)
gdfsa_af3 = gpd.GeoDataFrame(gdfsa_af3, geometry = gpd.points_from_xy(gdfsa_af3['x'], gdfsa_af3['y']), crs=4326)
gdfna3 = gpd.GeoDataFrame(gdfna3, geometry = gpd.points_from_xy(gdfna3['x'], gdfna3['y']), crs=4326)
```

```python
gdfsa_am3[gdfsa_am3.matrix == 1]
```

```python
bird_SA_am
```

```python
# Remove the bins with NaNs (the ones without any datapoints, cannot generate a distance or RMSE)
# Perhaps replace them with zeros instead of removing them?
gdfns3 = gdfns3[gdfns3.lat_bins.astype(str) != '(60, 70]']
gdfsa_am3 = gdfsa_am3[gdfsa_am3.lat_bins.astype(str) != '(-80, -70]']
gdfsa_af3 = gdfsa_af3[gdfsa_af3.lat_bins.astype(str) != '(-80, -70]']
gdfna3 = gdfna3[(gdfna3.lat_bins.astype(str) != '(-80, -70]') & (gdfna3.lat_bins.astype(str) != '(-70, -60]') & (gdfna3.lat_bins.astype(str) != '(-60, -50]') & (gdfna3.lat_bins.astype(str) != '(60, 70]')]
bird_NA = bird_NA[bird_NA.lat_bins_10.astype(str) != '(-70, -60]']
```

```python
# For each matrix (path), calcuate the distance between its points and the reference (bird tracking data). 
dist_SS = np.array([])
for m in gdfss3.matrix.unique():
    for i in range(15):
        dist_SS = np.append(dist_SS, haversine((bird_SS.lat_median10.values[i], gdfss3[gdfss3.matrix == m].x.values[i]), 
              (bird_SS.lat_median10.values[i], bird_SS.lon_median10.values[i])))
```

```python
dist_NS = np.array([])
for m in gdfns3.matrix.unique():
    for i in range(14):
        dist_NS = np.append(dist_NS, haversine((bird_NS.lat_median10.values[i], gdfns3[gdfns3.matrix == m].x.values[i]), 
              (bird_NS.lat_median10.values[i], bird_NS.lon_median10.values[i])))
```

```python
dist_SA_am = np.array([])
for m in gdfsa_am3.matrix.unique():
    for i in range(14):
        dist_SA_am = np.append(dist_SA_am, haversine((bird_SA_am.lat_median10.values[i], 
                                                      gdfsa_am3[gdfsa_am3.matrix == m].x.values[i]), 
              (bird_SA_am.lat_median10.values[i], bird_SA_am.lon_median10.values[i])))
```

```python
dist_SA_af = np.array([])
for m in gdfsa_af3.matrix.unique():
    for i in range(14):
        dist_SA_af = np.append(dist_SA_af, haversine((bird_SA_af.lat_median10.values[i], gdfsa_af3[gdfsa_af3.matrix == m].x.values[i]), 
              (bird_SA_af.lat_median10.values[i], bird_SA_af.lon_median10.values[i])))
```

```python
dist_NA = np.array([])
for m in gdfna3.matrix.unique():
    for i in range(11):
        dist_NA = np.append(dist_NA, haversine((bird_NA.iloc[:, :].lat_median10.values[i], 
                                                gdfna3[gdfna3.matrix == m].x.values[i]), 
                                               (bird_NA.iloc[:, :].lat_median10.values[i], 
                                                bird_NA.iloc[:, :].lon_median10.values[i])))
```

```python
# Add the distance as a new column ('error')
gdfss3['error'] = dist_SS
gdfns3['error'] = dist_NS
gdfsa_am3['error'] = dist_SA_am
gdfsa_af3['error'] = dist_SA_af
gdfna3['error'] = dist_NA
```

```python
# For each path (matrix), calculate the RMSE (square root of the mean of the squared errors)
# RMSE will be our main metric for finding the best path
rmse_ss = np.array([])
for m in gdfss3.matrix.unique():
    rmse_ss = np.append(rmse_ss, np.sqrt(np.mean((gdfss3[gdfss3.matrix == m].error**2))))
```

```python
rmse_ns = np.array([])
for m in gdfns3.matrix.unique():
    rmse_ns = np.append(rmse_ns, np.sqrt(np.mean((gdfns3[gdfns3.matrix == m].error**2))))
```

```python
rmse_sa_am = np.array([])
for m in gdfsa_am3.matrix.unique():
    rmse_sa_am = np.append(rmse_sa_am, np.sqrt(np.mean((gdfsa_am3[gdfsa_am3.matrix == m].error**2))))
```

```python
rmse_sa_af = np.array([])
for m in gdfsa_af3.matrix.unique():
    rmse_sa_af = np.append(rmse_sa_af, np.sqrt(np.mean((gdfsa_af3[gdfsa_af3.matrix == m].error**2))))
```

```python
rmse_na = np.array([])
for m in gdfna3.matrix.unique():
    rmse_na = np.append(rmse_na, np.sqrt(np.mean((gdfna3[gdfna3.matrix == m].error**2))))
```

```python
gdfss2['rmse'] = rmse_ss
gdfns2['rmse'] = rmse_ns
gdfsa_am2['rmse'] = rmse_sa_am
gdfsa_af2['rmse'] = rmse_sa_af
gdfna2['rmse'] = rmse_na
```

```python
def hist_plot_rmse(data, pop, season):
    data.rmse.plot.hist(histtype = 'step', label = 'total')
    data[data.d >= 0.3].rmse.plot.hist(alpha = 0.4, label = 'd')
    data[data.b >= 0.3].rmse.plot.hist(alpha = 0.4, label = 'b')
    data[data.a >= 0.6].rmse.plot.hist(alpha = 0.7, label = 'a')
    data[data.c >= 0.6].rmse.plot.hist(alpha = 0.7, label = 'c')
    #plt.xlim(350, 3200)
    plt.legend()
    plt.title(str(pop) + " " + str(season))
    plt.show()
```

```python
def line_plot_rmse(data, data2, pop, season):
    fig, ax = plt.subplots(ncols = 5, figsize = (22, 7))#, subplot_kw={'projection': ccrs.PlateCarree()})
    data.plot(ax = ax[0], column = 'a', vmin = 0, vmax = 1, alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    ax[0].plot(data2.lon_median10, 
         data2.lat_median10, zorder = 3, linewidth = 3, color = 'green', linestyle = '--')
    data.plot(ax = ax[1], column = 'b', vmin = 0, vmax = 0.5, alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data.plot(ax = ax[2], column = 'c', vmin = 0, vmax = 1, alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data.plot(ax = ax[3], column = 'd', vmin = 0, vmax = 0.5, alpha = 0.3, legend = True, linewidth = 2, cmap = 'RdYlBu_r')
    data.plot(ax = ax[4], column = 'rmse', vmin = None, vmax = None, alpha = 0.3, legend = True, linewidth = 2, cmap = 'PiYG_r')
    #ax[0].coastlines()
    #ax[1].coastlines()
    #ax[2].coastlines()
    #ax[3].coastlines()
    #ax[4].coastlines()
    #ax[0].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[1].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[2].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[3].add_feature(cfeature.OCEAN, facecolor = 'grey')
    #ax[4].add_feature(cfeature.OCEAN, facecolor = 'grey')
    ax[0].set_title('a')
    ax[1].set_title('b')
    ax[2].set_title('c')
    ax[3].set_title('d')
    ax[4].set_title('RMSE')
    ax[4].legend()
    plt.suptitle(str(pop) + " " + str(season))
    plt.show()
```

```python
# Paths with low RMSE tend to have higher alpha coefficients, paths with high RMSE have high beta coefficient
hist_plot_rmse(gdfss2[gdfss2.rmse < 2500], 'Svalbard', 'spring')
```
