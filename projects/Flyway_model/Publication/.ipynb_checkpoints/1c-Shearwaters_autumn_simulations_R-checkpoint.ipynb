```r
# My formula
myCost05 <- function(wind.direction, wind.speed, target, type = "active", chla = chla$chlor_a, alpha, beta, gamma, delta) {
  
  ###################  
  # 1) Preprocessing:
  ###################
    
  # The difference between the wind direction and the bid's PDM
  dif <- abs(wind.direction - target)
  
  # Convert wind support to cost: Low wind support -> high cost, high wind support -> low cost
  # Method: 1) Subtract the 99th percentile from wind support values: Low support rasters now have large negative values, high support rasters are 
  # close to 0. 2) Take their absolute values: Low support rasters have now high positive values (high cost), high-support rasters are still close to 0
  # (low cost)
  costx <- abs(wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180) - quantile(wind.speed[!is.na(dif)] * cos(dif[!is.na(dif)] * pi/180), 0.99))
  cost_stand <- 100/(quantile(costx, probs = c(0.99)))
  #costx <- abs(wind.speed * cos(dif * pi/180) - quantile(wind.speed * cos(dif * pi/180), 0.99, na.rm = TRUE))

  crossx <- abs(wind.speed[!is.na(dif)] * sin(dif[!is.na(dif)] * pi/180))
  cross_stand <- 100/(quantile(crossx, probs = c(0.99)))
  # Chla is hevily skewed: Many zeros and a few high values. I take its logarithm, and rasters with > -0.5 (high chl a concentration) all get a 
  # (minimum) value of -0.5. Later on, in the actual formula (next step) I add 0.5 and take its absolute (like I did with support). Low-chl rasters
  # have high positive values (cost) while productive rasters are close to 0.
  log_chla <- as.matrix(log10(chla[!is.na(dif)]))
  log_chla[log_chla > -0.8] <- -0.8
  log_chla[is.na(log_chla)] <- 0
  log_chla_abs <- abs(log_chla + 0.8)
  log_chla_abs[is.infinite(log_chla_abs)] <- 0
  chla_stand <- 100/(quantile(log_chla_abs, probs = c(0.99)))
  # raster_dist <- raster(matrix(6, 180, 360),
  #            xmn = -180.5, xmx = 179.5,
  #            ymn = -90, ymx = 90, crs=CRS("+proj=longlat +datum=WGS84 +no_defs"))
    
     
  ################################  
  # 2) The 4 formulas (components)
  ################################
    
  # Each coefficient takes a value between 0 and 1. Their sum has to be always = 1.
  a <- alpha     # Wind support
  b <- beta      # Distance
  c <- gamma     # Food
  d <- delta     # Crosswind
    
  # The actual formulas: a) Support is already pre-processed. b) For distance, I take the median value of support (or 6, which is close to the median). 
  # All rasters are equal, so if a bird wants to follow the shortest great circle distance, it will follow a straight path with a b coefficient = 1. 
  # c) I "flip" the log-transformed chl a as explained in the previous step. d) Crosswind component. If a bird wants to follow a path with low 
  # crosswinds (usually close to the coast), then a large d coefficient is necessary.
  dif[!is.na(dif)] <-  a * cost_stand * costx + 
                      # b * raster_dist[!is.na(raster_dist)] #+
                       b * median(cost_stand * costx) + #(wind.speed[!is.na(dif)] * 0 + 6) +
                       c * chla_stand * log_chla_abs +
                       d * cross_stand * crossx
    
  # Land rasters have a cost of 1000000. Arctic terns should not fly over land (in reality they do fly shortly over the UK, as far as I know).  
  dif[is.na(dif)] <- 1000000
#  dif[is.na(dif)] <- 0

  dif  
}
```

```r
# List of all possible coefficient combinations: each coefficients changes by 0.1 at each step & teir sum needs to be always 1
# This gives me 195 coefficient combinations -> 195 possible ways of combining the 4 cost surfaces -> 195 simulations -> 195 potantial flyways to choose 
# from (validation step in the end, using the bird tracking data)
s = 0
coefflist = c()
for (a in seq(0, 1, 0.2))
{
    for (b in seq(0, 1, 0.2))
    {
        for (c in seq(0, 0.8, 0.2))
        {
            for (d in seq(0, 0.8, 0.2))
            {if (a + b + c + d == 1)
                {s = s + 1
                coefflist = c(coefflist, c(a, b, c, d))}
            }
        }
    }    
}
s
```

```r
# List of all 195 conductance surfaces. This step is the slowest, takes ~5min on my Mac Pro
mylist <- c()
for (a in seq(0, 1, 0.2))
{
    for (b in seq(0, 1, 0.2))
    {
        for (c in seq(0, 0.8, 0.2))
        {
            for (d in seq(0, 0.8, 0.2))
            {if (a + b + c + d == 1)
                {mylist <- c(mylist, flow.dispersion(my_raster, fun = myCost05, output = "transitionLayer", type = "active", chla = chla$chlor_a, a, b, c, d))}
            }
        }
    }    
}
```

```r
# Geocorrecting (weights based on pixel area)
mylistC <- c()
for (surf in mylist)
    {
    mylistC <- c(mylistC, geoCorrection(surf))
    }
```
