```python
# Create a custom function for calculating bird and wind angles
def nomi_angles(x, y):
    angles = np.where(((x >= 0) & (y >= 0)) | ((x >= 0) & (y < 0)), 270 - np.degrees(np.arctan(y/x)), 90 - np.degrees(np.arctan(y/x)))
    return angles
```

```python
# Create a function fo finding theta: the angle between the bird and the wind direction
def mytheta(bird_angle, wind_angle):
    theta = bird_angle - wind_angle
    mytheta = theta.apply(lambda x: 360 + x if x < -180 else (360 - x if x > 180 else np.abs(x)))
    return mytheta
```
