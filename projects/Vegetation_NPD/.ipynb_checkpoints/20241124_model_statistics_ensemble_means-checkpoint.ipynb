```python
fig, ax = plt.subplots(ncols = 1, nrows = 1, figsize = (9, 8))


for i, m in enumerate(dsn5.model):
    ax.scatter(dsh.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']), 
               i, alpha = 1, s = df[df.model == m].model_weight_norm*1500, c = 'pink', edgecolor = 'k', zorder = 3)
for i, m in enumerate(dsn5.model):
    ax.scatter(dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']), 
               i, alpha = 1, s = df[df.model == m].model_weight_norm*1500, c = 'red', edgecolor = 'k', zorder = 3)
for i, m in enumerate(dsn5.model):
    ax.scatter(dsf5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']), 
               i, alpha = 1, s = df[df.model == m].model_weight_norm*1500, c = 'maroon', edgecolor = 'k', zorder = 3)

for i, m in enumerate(dsn5.model):
    ax.hlines(i, dsh.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']),
              dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']),
              color = 'k', zorder = 2)
for i, m in enumerate(dsn5.model):
    ax.hlines(i, dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']),
              dsf5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']),
              color = 'k', zorder = 2)
for i, m in enumerate(dsn5.model):
    ax.hlines(i, 130, 190,
              color = 'k', linestyle = '--', zorder = 2, alpha = 0.2)

for i, m in enumerate(dsn5.model):
    ax.text((dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']) + dsh.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']))/2 - 1, 
            i - 0.2, 
            int(np.round(dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']).values - dsh.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']).values, 0)),
              color = 'k', zorder = 4)
for i, m in enumerate(dsn5.model):
    ax.text((dsf5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']) + dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']))/2 - 1.2, 
            i - 0.2, 
            int(np.round(dsf5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']).values - dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat', 'time']).values, 0)),
              color = 'k', zorder = 4)
    
ax.vlines(150, 0, 0, color = 'pink', linewidth = 7, label = 'historical')
ax.vlines(150, 0, 0, color = 'red', linewidth = 7, label = 'near future')
ax.vlines(150, 0, 0, color = 'maroon', linewidth = 7, label = 'far future')
ax.scatter(dsf5.doy.sel(model = 'EC-Earth3-CC').weighted(weights).mean(['lon', 'lat', 'time']),
               3, alpha = 0.4, s = 100, c = 'red', zorder = 0, label = 'annual values')
ax.scatter(dsf5.doy.sel(model = 'EC-Earth3-CC').weighted(weights).mean(['lon', 'lat', 'time']),
               3, alpha = 1, s = 100, c = 'red', edgecolor = 'k', zorder = 0, label = 'time period mean')
ax.vlines(150, 0, 0, color = 'red', linestyle = '-', linewidth = 4, label = '90% weighted CI')


ax.set_yticks(ticks = range(len(dsf5.model.values)), labels = dsf5.model.values)
ax.set_xlabel('DOY', fontsize = 14)
ax.set_ylabel('Models', fontsize = 14)
ax.set_xlim(130, 190)
#ax.set_ylim(-0.4, 6.8)
ax.legend(bbox_to_anchor=(1, 1))
plt.suptitle('Arctic mean, SSP-5.85', fontsize = 16, y = 1.0)
plt.tight_layout()
```
