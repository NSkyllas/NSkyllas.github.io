```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
dsm = xr.open_dataset('DOY/doy_ensmean_da_hist.nc')
dss = xr.open_dataset('DOY/doy_ensstd_da_hist.nc')
```

```python
for f in ['dsm', 'dss']:
    exec(f"{f}['doy'] = {f}['__xarray_dataarray_variable__']")
    exec(f"{f} = {f}.drop(['__xarray_dataarray_variable__'])")
```

```python
dsm
```

```python
fig, ax = plt.subplots(ncols = 5, nrows = 3, figsize = (18, 9), subplot_kw={'projection': ccrs.NorthPolarStereo()})
ax = ax.flatten()
for i in range(len(dsm.time)):
    dsm.isel(time = i).doy.plot(ax = ax[i], vmin = 120, vmax = 205, cmap = 'terrain', transform = ccrs.PlateCarree())
    ax[i].coastlines()
plt.suptitle('DOY historical, ensemble mean')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(ncols = 5, nrows = 3, figsize = (18, 9), subplot_kw={'projection': ccrs.NorthPolarStereo()})
ax = ax.flatten()
for i in range(len(dss.time)):
    dss.isel(time = i).doy.plot(ax = ax[i], vmin = 0, vmax = 30, transform = ccrs.PlateCarree())
    ax[i].coastlines()
plt.suptitle('DOY historical, model spread (ensemble stdev)')
plt.tight_layout()
plt.show()
```
