```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
```

```python
dsm_ff = xr.open_dataset('DOY/doy_ensmean_da_ssp126_ff.nc')
dsm_nf = xr.open_dataset('DOY/doy_ensmean_da_ssp126_nf.nc')
dss_ff = xr.open_dataset('DOY/doy_ensstd_da_ssp126_ff.nc')
dss_nf = xr.open_dataset('DOY/doy_ensstd_da_ssp126_nf.nc')
```

```python
for f in ['dsm_nf', 'dsm_ff', 'dss_nf', 'dss_ff']:
    exec(f"{f}['doy'] = {f}['__xarray_dataarray_variable__']")
    exec(f"{f} = {f}.drop(['__xarray_dataarray_variable__'])")
```

```python
dsm_nf
```

```python
fig, ax = plt.subplots(ncols = 5, nrows = 4, figsize = (18, 9))
ax = ax.flatten()
for i in range(len(dsm_nf.time)):
    dsm_nf.isel(time = i).doy.plot(ax = ax[i], vmin = 140, vmax = 205, cmap = 'BrBG_r')
plt.suptitle('DOY ssp126, near-future ensemble mean')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(ncols = 5, nrows = 4, figsize = (18, 9))
ax = ax.flatten()
for i in range(len(dss_nf.time)):
    dss_nf.isel(time = i).doy.plot(ax = ax[i], vmin = 0, vmax = 25)
plt.suptitle('DOY ssp126, near-future model spread (ensemble stdev)')
plt.tight_layout()
plt.show()
```
