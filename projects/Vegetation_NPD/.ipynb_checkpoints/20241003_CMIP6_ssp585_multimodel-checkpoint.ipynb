```python
def nomi_rmse(y_pred, y):
    '''my custom RMSE function'''
    return np.sqrt(np.nanmean((y_pred - y)**2))
```

I found that normalising both the LAI and NDVI works best. Thatt's why I have to normalise (between 0-100) the models' LAIs before plugging them to my algorithm.

```python
# For each available year I find the min & max per pixel, and convert NDVI to percentage
for i in range(2, 12):
    for yr in range(2040, 2060):
        exec(f"maxim{i} = ds{i}s_nf.lai.sel(time = ds{i}s_nf.time.dt.year == yr).max(dim = 'time')")
        exec(f"minim{i} = ds{i}s_nf.lai.sel(time = ds{i}s_nf.time.dt.year == yr).min(dim = 'time')")
        exec(f"value{i} = ds{i}s_nf.lai.sel(time = ds{i}s_nf.time.dt.year == yr)")
        exec(f"lai{i}_perc_{yr} = 100*((value{i} - minim{i})/(maxim{i} - minim{i}))")
```

```python
# For each available year I find the min & max per pixel, and convert NDVI to percentage
for i in range(2, 12):
    for yr in range(2080, 2100):
        exec(f"maxim{i} = ds{i}s_ff.lai.sel(time = ds{i}s_ff.time.dt.year == yr).max(dim = 'time')")
        exec(f"minim{i} = ds{i}s_ff.lai.sel(time = ds{i}s_ff.time.dt.year == yr).min(dim = 'time')")
        exec(f"value{i} = ds{i}s_ff.lai.sel(time = ds{i}s_ff.time.dt.year == yr)")
        exec(f"lai{i}_perc_{yr} = 100*((value{i} - minim{i})/(maxim{i} - minim{i}))")
```

#### Let's try to recreate the following figure as a validation step for my 3 models

```python
def nomi_poly_simple3(x, a, b, c, d):
    '''for recreating (predicting) the y (NDVI) using the polynomial coefficients of each pixel and the x (LAI)'''
    y_predicted = a*x**3 + b*x**2 + c*x + d
    return y_predicted
```

## Near Future calculations (2040 - 2060)

```python
%%time
#Create the testing dataset, containing 5 of the 15 years (the ones not used during training), and test the algorithm on it 5 times (one for each year)
nan_threshold = 18
year_test = list(range(2040, 2060))
#year_test = [ele for ele in year_test if ele not in year_train]
# For each of the testing years
for m in range(2, 12):
    for yr in range(20):
        print(m, yr)
        # Create the feature (satellite LAI) by selecting 1 of the 14 years and putting its values in the array (each test year is independent now, and uses the same 
        # polynomial that was trained on the 10 training years) 
        exec(f"mod{m}_lai_{yr+1} = da{m}s_nf.sel(time = da{m}s_nf.time.dt.year.isin(year_test[{yr}])).values")
        # Create an array to store the satellite NDVI predictions, using the polynomial coefficients of each pixel and the satellite LAI
        exec(f"mod{m}_ndvi_{yr+1}_nf = np.zeros((mod{m}_lai_{yr+1}.shape[0], mod{m}_lai_{yr+1}.shape[1], mod{m}_lai_{yr+1}.shape[2]))")
        # For all lons & lats (all pixels)
        for j in eval(f"range(mod{m}_lai_{yr+1}.shape[1])"):
                for i in eval(f"range(mod{m}_lai_{yr+1}.shape[2])"):
                    # If the pixel has only few missing datapoints (right now it's <18, WHY?)
                    if eval(f"(np.sum(np.isnan(mod{m}_lai_{yr+1}[:, j, i])) < nan_threshold) & (np.sum(np.isnan(mod{m}_lai_{yr+1}[:, j, i])) == np.sum(np.isnan(mod{m}_ndvi_{yr+1}_nf[:, j, i])))"):        
                        # Select the LAI timeseries of that pixel
                        exec(f"mod{m}_lai = mod{m}_lai_{yr+1}[:, j, i]")
                        # Select the test NDVI timeseries of that pixel
                        exec(f"mod{m}_ndvi = mod{m}_ndvi_{yr+1}_nf[:, j, i]")
                        # Use the polynomial model (coefficients) for that pixel to predict NDVI 
                        exec(f"mod{m}_ndvi_predicted = nomi_poly_simple3(mod{m}_lai, dsp.sel(coef = 'a').coefficient.values[j, i], dsp.sel(coef = 'b').coefficient.values[j, i], dsp.sel(coef = 'c').coefficient.values[j, i], dsp.sel(coef = 'interc').coefficient.values[j, i])")
                        # Standardise again (2nd time) the NDVI between 0 and 100, because the model produces results with lower range than 100. 
                        exec(f"mod{m}_ndvi_{yr+1}_nf[:, {j}, {i}] = 100*(mod{m}_ndvi_predicted - np.nanmin(mod{m}_ndvi_predicted))/(np.nanmax(mod{m}_ndvi_predicted) - np.nanmin(mod{m}_ndvi_predicted))")
                        #exec(f"y_pred_{yr+1}[{j}, {i}, :] = y_predicted")
                        #exec(f"rmse_{y}[j, i] = nomi_rmse(y_predicted, y_test)")
                        #exec(f"corr_{y}[j, i] = np.corrcoef(y_predicted, y_test)[0][1]")
                    else:
                        #print("No match")
                        exec(f"mod{m}_ndvi_{yr+1}_nf[:, {j}, {i}] = np.nan")
                        #exec(f"rmse_{y}[j, i] = np.nan")
                        #exec(f"corr_{y}[j, i] = np.nan")
```

```python
%%time
# For each test year
threshold = 90
for m in range(2, 12):
    for yr in range(1, 20):
        print(m, yr)
        # Create 2-d arrays to store the 50% DOY according to the test NDVI and to the predicted NDVI
        exec(f"doy_mod{m}_{yr}_nf = np.zeros((mod{m}_ndvi_{yr}_nf.shape[1], mod{m}_ndvi_{yr}_nf.shape[2]))")
        # For each pixel
        for j in eval(f"range(doy_mod{m}_{yr}_nf.shape[0])"):
            for i in eval(f"range(doy_mod{m}_{yr}_nf.shape[1])"):
                # DOY is the day that NDVI goes above 50%
                exec(f"doy_mod{m}_{yr}_nf[j, i] = np.argmax(np.where(mod{m}_ndvi_{yr}_nf[:, j, i] > 50, 1, 0))")     
                # Although my predictions are smooth (because they come from the smooth LAI), the satellite NDVI (test NDVI) has weird peaks, especially early in the season.
                # That's why I ignore the first 90 days.
                exec(f"doy_mod{m}_{yr}_nf[j, i] = np.argmax(np.where(mod{m}_ndvi_{yr}_nf[threshold:, j, i] > 50, 1, 0)) + threshold")   
                exec(f"doy_mod{m}_{yr}_nf[doy_mod{m}_{yr}_nf == 90] = np.nan")            
```

## Far Future calculations (2080 - 2100)

```python
%%time
#Create the testing dataset, containing 5 of the 15 years (the ones not used during training), and test the algorithm on it 5 times (one for each year)
nan_threshold = 18
year_test = list(range(2080, 2100))
#year_test = [ele for ele in year_test if ele not in year_train]
# For each of the testing years
for m in range(2, 12):
    for yr in range(20):
        print(m, yr)
        # Create the feature (satellite LAI) by selecting 1 of the 14 years and putting its values in the array (each test year is independent now, and uses the same 
        # polynomial that was trained on the 10 training years) 
        exec(f"mod{m}_lai_{yr+1} = da{m}s_ff.sel(time = da{m}s_ff.time.dt.year.isin(year_test[{yr}])).values")
        # Create an array to store the satellite NDVI predictions, using the polynomial coefficients of each pixel and the satellite LAI
        exec(f"mod{m}_ndvi_{yr+1}_ff = np.zeros((mod{m}_lai_{yr+1}.shape[0], mod{m}_lai_{yr+1}.shape[1], mod{m}_lai_{yr+1}.shape[2]))")
        # For all lons & lats (all pixels)
        for j in eval(f"range(mod{m}_lai_{yr+1}.shape[1])"):
                for i in eval(f"range(mod{m}_lai_{yr+1}.shape[2])"):
                    # If the pixel has only few missing datapoints (right now it's <18, WHY?)
                    if eval(f"(np.sum(np.isnan(mod{m}_lai_{yr+1}[:, j, i])) < nan_threshold) & (np.sum(np.isnan(mod{m}_lai_{yr+1}[:, j, i])) == np.sum(np.isnan(mod{m}_ndvi_{yr+1}_ff[:, j, i])))"):        
                        # Select the LAI timeseries of that pixel
                        exec(f"mod{m}_lai = mod{m}_lai_{yr+1}[:, j, i]")
                        # Select the test NDVI timeseries of that pixel
                        exec(f"mod{m}_ndvi = mod{m}_ndvi_{yr+1}_ff[:, j, i]")
                        # Use the polynomial model (coefficients) for that pixel to predict NDVI 
                        exec(f"mod{m}_ndvi_predicted = nomi_poly_simple3(mod{m}_lai, dsp.sel(coef = 'a').coefficient.values[j, i], dsp.sel(coef = 'b').coefficient.values[j, i], dsp.sel(coef = 'c').coefficient.values[j, i], dsp.sel(coef = 'interc').coefficient.values[j, i])")
                        # Standardise again (2nd time) the NDVI between 0 and 100, because the model produces results with lower range than 100. 
                        exec(f"mod{m}_ndvi_{yr+1}_ff[:, {j}, {i}] = 100*(mod{m}_ndvi_predicted - np.nanmin(mod{m}_ndvi_predicted))/(np.nanmax(mod{m}_ndvi_predicted) - np.nanmin(mod{m}_ndvi_predicted))")
                        #exec(f"y_pred_{yr+1}[{j}, {i}, :] = y_predicted")
                        #exec(f"rmse_{y}[j, i] = nomi_rmse(y_predicted, y_test)")
                        #exec(f"corr_{y}[j, i] = np.corrcoef(y_predicted, y_test)[0][1]")
                    else:
                        #print("No match")
                        exec(f"mod{m}_ndvi_{yr+1}_ff[:, {j}, {i}] = np.nan")
                        #exec(f"rmse_{y}[j, i] = np.nan")
                        #exec(f"corr_{y}[j, i] = np.nan")
```

```python
%%time
# For each test year
threshold = 90
for m in range(2, 12):
    for yr in range(1, 20):
        print(m, yr)
        # Create 2-d arrays to store the 50% DOY according to the test NDVI and to the predicted NDVI
        exec(f"doy_mod{m}_{yr}_ff = np.zeros((mod{m}_ndvi_{yr}_ff.shape[1], mod{m}_ndvi_{yr}_ff.shape[2]))")
        # For each pixel
        for j in eval(f"range(doy_mod{m}_{yr}_ff.shape[0])"):
            for i in eval(f"range(doy_mod{m}_{yr}_ff.shape[1])"):
                # DOY is the day that NDVI goes above 50%
                exec(f"doy_mod{m}_{yr}_ff[j, i] = np.argmax(np.where(mod{m}_ndvi_{yr}_ff[:, j, i] > 50, 1, 0))")     
                # Although my predictions are smooth (because they come from the smooth LAI), the satellite NDVI (test NDVI) has weird peaks, especially early in the season.
                # That's why I ignore the first 90 days.
                exec(f"doy_mod{m}_{yr}_ff[j, i] = np.argmax(np.where(mod{m}_ndvi_{yr}_ff[threshold:, j, i] > 50, 1, 0)) + threshold")   
                exec(f"doy_mod{m}_{yr}_ff[doy_mod{m}_{yr}_ff == 90] = np.nan")            
```
