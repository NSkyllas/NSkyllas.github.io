```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import pandas as pd
import seaborn as sns
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
dsn5 = xr.open_dataset('DOY/doy_mod_ssp585_nf.nc')
dsf5 = xr.open_dataset('DOY/doy_mod_ssp585_ff.nc')
dsh = xr.open_dataset('DOY/doy_mod_hist.nc')
```

```python
dsn2 = xr.open_dataset('DOY/doy_mod_ssp245_nf.nc')
dsf2 = xr.open_dataset('DOY/doy_mod_ssp245_ff.nc')
```

```python
dsn1 = xr.open_dataset('DOY/doy_mod_ssp126_nf.nc')
dsf1 = xr.open_dataset('DOY/doy_mod_ssp126_ff.nc')
```

```python
dss = xr.open_dataset('DOY/doy_sat.nc')
```

```python
for f in ['dsn5', 'dsf5', 'dsn2', 'dsf2', 'dsn1', 'dsf1', 'dsh', 'dss']:
    exec(f"{f}['doy'] = {f}['__xarray_dataarray_variable__']")
    exec(f"{f} = {f}.drop(['__xarray_dataarray_variable__'])")
```

```python
weights = np.cos(np.deg2rad(dss.lat))
weights.name = "weights"
```

```python
fig, ax = plt.subplots(ncols = 3, nrows = 1, figsize = (10, 3))
dsh.doy.weighted(weights).mean(['lon', 'lat']).max('model').plot(ax = ax[0], label = 'maximum')
dsh.doy.weighted(weights).mean(['lon', 'lat']).mean('model').plot(ax = ax[0], label = 'mean')
dsh.doy.weighted(weights).mean(['lon', 'lat']).median('model').plot(ax = ax[0], label = 'median')
dsh.doy.weighted(weights).mean(['lon', 'lat']).min('model').plot(ax = ax[0], label = 'minimum')
ax[0].set_ylim(130, 190)
ax[0].set_title('historical (2000-2013)')
ax[0].legend()
ax[0].set_xticks(range(2000, 2014, 5))

dsn5.doy.weighted(weights).mean(['lon', 'lat']).max('model').plot(ax = ax[1])
dsn5.doy.weighted(weights).mean(['lon', 'lat']).mean('model').plot(ax = ax[1])
dsn5.doy.weighted(weights).mean(['lon', 'lat']).median('model').plot(ax = ax[1])
dsn5.doy.weighted(weights).mean(['lon', 'lat']).min('model').plot(ax = ax[1])
ax[1].set_ylim(130, 190)
ax[1].set_title('ssp585 (2040-2059)')

dsf5.doy.weighted(weights).mean(['lon', 'lat']).max('model').plot(ax = ax[2])
dsf5.doy.weighted(weights).mean(['lon', 'lat']).mean('model').plot(ax = ax[2])
dsf5.doy.weighted(weights).mean(['lon', 'lat']).median('model').plot(ax = ax[2])
dsf5.doy.weighted(weights).mean(['lon', 'lat']).min('model').plot(ax = ax[2])
ax[2].set_ylim(130, 190)
ax[2].set_title('ssp585 (2080-2099)')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols = 3, nrows = 1, figsize = (10, 3))

ax[0].hist(dsh.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), density = True, bins = 30)
sns.kdeplot(dsh.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[0], bw_adjust = 0.5)
ax[0].set_xlim(130, 200)
ax[0].set_title('historical (2000-2013)')
ax[1].hist(dsn5.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), density = True, bins = 30)
sns.kdeplot(dsn5.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[1], bw_adjust = 0.5)
ax[1].set_xlim(130, 200)
ax[1].set_title('ssp585 (2040-2059)')
ax[2].hist(dsf5.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), density = True, bins = 30)
sns.kdeplot(dsf5.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[2], bw_adjust = 0.5)
ax[2].set_xlim(130, 200)
ax[2].set_title('ssp585 (2080-2099)')

plt.suptitle('Arctic mean, DOY values per year from all models')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols = 3, nrows = 1, figsize = (10, 3))

sns.swarmplot(dsh.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[0], label = 'All models')
sns.swarmplot(dsh.sel(model = 'CMCC-CM2-HR4').doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[0], label = 'CMCC-CM2-HR4', color = 'r')
sns.swarmplot(dsh.sel(model = 'ACCESS-ESM1-5').doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[0], label = 'ACCESS-ESM1-5', color = 'g')

sns.swarmplot(dsn5.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[1], label = 'All models')
#sns.swarmplot(dsn.sel(model = 'CMCC-CM2-HR4').doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[1], label = 'CMCC-CM2-HR4')
sns.swarmplot(dsn5.sel(model = 'ACCESS-ESM1-5').doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[1], color = 'g')

sns.swarmplot(dsf5.doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[2], label = 'All models')
#sns.swarmplot(dsf.sel(model = 'CMCC-CM2-HR4').doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[2], label = 'CMCC-CM2-HR4')
sns.swarmplot(dsf5.sel(model = 'ACCESS-ESM1-5').doy.weighted(weights).mean(['lon', 'lat']).values.flatten(), ax = ax[2], color = 'g')

ax[0].set_ylim(130, 190)
ax[0].set_title('ssp585 (2080-2099)')
ax[0].set_ylabel('DOY')
ax[1].set_ylim(130, 190)
ax[1].set_title('ssp585 (2080-2099)')
ax[2].set_ylim(130, 190)
ax[2].set_title('ssp585 (2080-2099)')
ax[0].legend()

plt.suptitle('Arctic mean, DOY values for all years from all models')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols = 3, nrows = 1, figsize = (18, 3))

dss.doy.sel(lon = slice(15, 30), lat = slice(60, 67)).weighted(weights).mean(['lon', 'lat']).plot(ax = ax[0], linestyle = '-', label = 'Satellite', color = 'k', linewidth = 2, zorder = 3)
dss.doy.sel(lon = slice(15, 30), lat = slice(60, 67)).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[0], color = 'k',zorder = 3)
for m in dsh.model:
    dsh.doy.sel(model = m, lon = slice(15, 30), lat = slice(60, 67)).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[0], alpha = 0.8)

for m in dsn5.model:
    dsn5.doy.sel(model = m, lon = slice(15, 30), lat = slice(60, 67)).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[1], alpha = 0.8)

for m in dsf5.model:
    dsf5.doy.sel(model = m, lon = slice(15, 30), lat = slice(60, 67)).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[2], alpha = 0.8)

plt.suptitle('Baltic Sea mean, NDVI 50% Day of Year')
ax[0].set_ylim(100, 190)
ax[0].set_title('historical (2000-2013)')
ax[0].set_ylabel('DOY')
ax[0].set_xlabel('Year')
ax[1].set_ylim(100, 190)
ax[1].set_title('ssp585 (2040-2059)')
ax[1].set_xlabel('Year')
ax[2].set_ylim(100, 190)
ax[2].set_title('ssp585 (2080-2099)')
ax[2].set_xlabel('Year')
ax[0].legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(ncols = 3, nrows = 1, figsize = (15, 3))

for m in dsh.model:
    dsh.doy.sel(model = m).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[0], alpha = 0.6)
ax[0].hlines(dsh.doy.weighted(weights).mean(['lon', 'lat']).quantile(0.05, ['model', 'time']).values, 0, 14, color = 'k', linestyle = '--')
ax[0].hlines(dsh.doy.weighted(weights).mean(['lon', 'lat']).quantile(0.95, ['model', 'time']).values, 0, 14, color = 'k', linestyle = '--')
ax[0].hlines(dsh.doy.weighted(weights).mean(['lon', 'lat', 'model', 'time']).values, 0, 14, color = 'k', linestyle = '-')
#ax[0].hlines(dss.doy.weighted(weights).mean(['lon', 'lat', 'time']).values, 0, 14, color = 'g', linestyle = '-')
for m in dsn5.model:
    dsn5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[1], alpha = 0.6)
ax[1].hlines(dsn5.doy.weighted(weights).mean(['lon', 'lat']).quantile(0.05, ['model', 'time']).values, 0, 14, color = 'k', linestyle = '--')
ax[1].hlines(dsn5.doy.weighted(weights).mean(['lon', 'lat']).quantile(0.95, ['model', 'time']).values, 0, 14, color = 'k', linestyle = '--')
ax[1].hlines(dsn5.doy.weighted(weights).mean(['lon', 'lat', 'model', 'time']).values, 0, 14, color = 'k', linestyle = '-')
for m in dsf5.model:
    dsf5.doy.sel(model = m).weighted(weights).mean(['lon', 'lat']).plot.scatter(ax = ax[2], alpha = 0.6)
ax[2].hlines(dsf5.doy.weighted(weights).mean(['lon', 'lat']).quantile(0.05, ['model', 'time']).values, 0, 14, color = 'k', linestyle = '--')
ax[2].hlines(dsf5.doy.weighted(weights).mean(['lon', 'lat']).quantile(0.95, ['model', 'time']).values, 0, 14, color = 'k', linestyle = '--', label = '90% CI')
ax[2].hlines(dsf5.doy.weighted(weights).mean(['lon', 'lat', 'model', 'time']).values, 0, 14, color = 'k', linestyle = '-', label = 'mean')

ax[0].set_ylim(120, 210)
ax[0].set_xlabel('Climate models')
ax[0].set_ylabel('Day of year')
ax[0].set_xticks(ticks = range(len(dsh.model.values)), labels = dsh.model.values, rotation = 90)
ax[1].set_ylim(120, 210)
ax[1].set_xlabel('Climate models')
ax[1].set_xticks(ticks = range(len(dsn5.model.values)), labels = dsn5.model.values, rotation = 90)
ax[2].set_ylim(120, 210)
ax[2].set_xlabel('Climate models')
ax[2].set_xticks(ticks = range(len(dsf5.model.values)), labels = dsf5.model.values, rotation = 90)
ax[0].set_title('historical (2000-2014)')
ax[1].set_title('near-future (2040-2060)')
ax[2].set_title('far-future (2080-2100)')
ax[2].legend()

plt.suptitle('Arctic mean, NDVI 50% Day of Year', y = 1.05)
```
