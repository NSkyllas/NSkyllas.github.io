```python
def nomi_poly(x, y, deg):
    poly_reg_model = LinearRegression()
    poly = PolynomialFeatures(degree = deg, include_bias=False)
    poly_features = poly.fit_transform(x.reshape(-1, 1))
    df_reg = pd.DataFrame(poly_features)
    df_reg['y'] = y
    df_reg = df_reg.sort_values(by = 0)
    poly_reg_model.fit(df_reg.iloc[:, :deg], df_reg['y'])
    y_predicted = poly_reg_model.predict(df_reg.iloc[:, :deg])
    return y_predicted, df_reg
```

```python
def nomi_poly_years(x, y, deg):
    poly_reg_model = LinearRegression()
    poly = PolynomialFeatures(degree = deg, include_bias=False)
    poly_features = poly.fit_transform(x.reshape(-1, 1))
    df_reg = pd.DataFrame(poly_features)
    df_reg['y'] = y
    df_reg = df_reg.sort_values(by = 0)
    poly_reg_model.fit(df_reg.iloc[:, :deg], df_reg['y'])
    y_predicted = poly_reg_model.predict(df_reg.iloc[:, :deg])
    return y_predicted, df_reg, poly_reg_model
```

```python
def nomi_poly_predict(x, y, deg, model):
    poly_reg_model = LinearRegression()
    poly = PolynomialFeatures(degree = deg, include_bias=False)
    poly_features = poly.fit_transform(x.reshape(-1, 1))
    df_reg = pd.DataFrame(poly_features)
    df_reg['y'] = y
    df_reg = df_reg.sort_values(by = 0)
    #poly_reg_model.fit(df_reg.iloc[:, :deg], df_reg['y'])
    y_predicted = model.predict(df_reg.iloc[:, :deg])
    return y_predicted, df_reg
```

```python
# Create the training dataset, containing 10 of the 15 years, and train the algorithm on it
for reg in ['b', 'w', 't']:
    for deg in range(3, 6):
        year_train = np.random.choice(range(1999, 2014), 10, replace = False)
        exec(f"x_train = ds1_{reg}s.sel(time = ds1_{reg}s.time.dt.year.isin(year_train)).values")
        exec(f"y_train = ds2_{reg}s.sel(time = ds2_{reg}s.time.dt.year.isin(year_train)).values")
        y_pred, df_train, poly_reg_model = nomi_poly_years(x_train, y_train, deg)
        # Create the testing dataset, containing 5 of the 15 years, and test the algorithm on it 5 times (one for each year)
        year_test = list(range(1999, 2014))
        year_test = [ele for ele in year_test if ele not in year_train]
        for i in range(5):
            exec(f"x_test_{reg}_{deg}_{i+1} = ds1_{reg}s.sel(time = ds1_{reg}s.time.dt.year.isin(year_test[{i}])).values")
            exec(f"y_test_{reg}_{deg}_{i+1} = ds2_{reg}s.sel(time = ds2_{reg}s.time.dt.year.isin(year_test[{i}])).values")
            exec(f"y_pred_{reg}_{deg}_{i+1}, df_test_{reg}_{i+1} = nomi_poly_predict(x_test_{reg}_{deg}_{i+1}, y_test_{reg}_{deg}_{i+1}, {deg}, poly_reg_model)")
```
