```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
ds1 = xr.open_dataset('ESA_satellite/LAI_1999-2020_V2.0.1.nc_LAI_reg_arctic.nc')
ds2 = xr.open_dataset('ESA_satellite/LAI_1999-2020_V2.0.1.nc_LAI_reg_arctic_daily.nc')
```

```python
ds3 = xr.open_dataset('ESA_satellite/NDVI_1999-2020_GLOBE_PROBAV_V3.0.1.nc_NDVI_reg_arctic.nc')
ds4 = xr.open_dataset('ESA_satellite/NDVI_1999-2020_GLOBE_PROBAV_V3.0.1.nc_NDVI_reg_arctic_daily.nc')
```

```python
ds1 = ds1.sel(lat = slice(60, 80))
ds2 = ds2.sel(lat = slice(60, 80))
ds3 = ds3.sel(lat = slice(60, 80))
ds4 = ds4.sel(lat = slice(60, 80))
```

```python
ds1
```

```python
ds3.isel(time = slice(1, 769))
```

```python
weights = np.cos(np.deg2rad(ds1.lat))
weights.name = "weights"
```

```python
fig, ax = plt.subplots(ncols = 4, figsize = (14, 3))
ds1.LAI.mean('time').plot(ax = ax[0])
ds2.LAI.mean('time').plot(ax = ax[1])
ds3.NDVI.mean('time').plot(ax = ax[2], vmin = 0)
ds4.NDVI.mean('time').plot(ax = ax[3], vmin = 0)
ax[0].set_title('LAI mean 1999-2020, 10-day')
ax[1].set_title('LAI mean 1999-2020, daily')
ax[2].set_title('NDVI mean 1999-2020, 10-day')
ax[3].set_title('NDVI mean 1999-2020, daily')

plt.tight_layout()
```

```python
fig, ax = plt.subplots(ncols = 4, figsize = (14, 3))
ds1.LAI.std('time').plot(ax = ax[0])
ds2.LAI.std('time').plot(ax = ax[1])
ds3.NDVI.std('time').plot(ax = ax[2], vmin = 0)
ds4.NDVI.std('time').plot(ax = ax[3], vmin = 0)
ax[0].set_title('LAI STD 1999-2020, 10-day')
ax[1].set_title('LAI STD 1999-2020, daily')
ax[2].set_title('NDVI STD 1999-2020, 10-day')
ax[3].set_title('NDVI STD 1999-2020, daily')

plt.tight_layout()
```

```python
ds1.LAI.weighted(weights).mean(['lon', 'lat']).plot()
ds2.LAI.weighted(weights).mean(['lon', 'lat']).plot()
```

```python
ds3.NDVI.weighted(weights).mean(['lon', 'lat']).plot()
ds4.NDVI.weighted(weights).mean(['lon', 'lat']).plot()
```

```python
plt.scatter(ds1.LAI.weighted(weights).mean(['lon', 'lat']),
           ds3.isel(time = slice(1, 769)).NDVI.weighted(weights).mean(['lon', 'lat']),
           c = ds1.time.dt.dayofyear)
plt.colorbar()
plt.xlabel('LAI')
plt.ylabel('NDVI')
plt.title('Arctic mean LAI vs NDVI')
```

```python
plt.scatter(ds1.LAI,
           ds3.isel(time = slice(1, 769)).NDVI, 
           alpha = 0.01, s = 1)
plt.xlabel('LAI')
plt.ylabel('NDVI')
plt.title('Point-wise LAI vs NDVI')
```
