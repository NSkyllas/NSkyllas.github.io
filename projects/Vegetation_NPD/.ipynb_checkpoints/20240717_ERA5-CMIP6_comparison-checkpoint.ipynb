```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
```

```python
ds1 = xr.open_dataset('ESA_satellite/LAI_1999-2020_V2.0.1.nc_LAI_reg_arctic.nc')
ds2 = xr.open_dataset('CMIP6/lai_Lmon_EC-Earth3-Veg_historical_r1i1p1f1_gr_185001-201412.ncreg.nc_15y.nc_arctic.nc').isel(bnds = 0)
ds3 = xr.open_dataset('CMIP6/lai_Lmon_CESM2_historical_r1i1p1f1_gn_185001-201412.ncreg.nc_15y.nc_arctic.nc').isel(bnds = 0)
ds4 = xr.open_dataset('CMIP6/lai_Lmon_ACCESS-ESM1-5_historical_r1i1p1f1_gn_185001-201412.ncreg.nc_15y.nc_arctic.nc').isel(bnds = 0)
```

```python
ds2 = ds2.where(ds3.isel(time = 0).lai.notnull())
ds1 = ds1.where(ds3.isel(time = 0).lai.notnull())
ds4 = ds4.where(ds3.isel(time = 0).lai.notnull())
```

```python
weights = np.cos(np.deg2rad(ds1.lat))
weights.name = "weights"
```

```python
ds2.lai.weighted(weights).mean(['lon', 'lat']).plot(label = 'ECE-Veg')
ds4.lai.weighted(weights).mean(['lon', 'lat']).plot(label = 'ACCESS')
plt.plot(ds4.time, ds3.lai.weighted(weights).mean(['lon', 'lat']).values, label = 'CESM2')

#(ds1.lai_lv + ds1.lai_hv).weighted(weights).mean(['lon', 'lat']).plot(linewidth = 3, label = 'ERA5 sum')
#ds1.lai_lv.weighted(weights).mean(['lon', 'lat']).plot(linewidth = 3, label = 'low')
#ds1.lai_hv.weighted(weights).mean(['lon', 'lat']).plot(linewidth = 3, label = 'high')
ds1.lai_w.weighted(weights).mean(['lon', 'lat']).plot(linewidth = 3, label = 'weighted sum')

plt.legend()
```

```python
ds1m = ds1.LAI.resample(time = '1M').mean()
```

```python
ds1m.time[12:192]
```

```python
ds2.lai.weighted(weights).mean(['lon', 'lat']).plot(label = 'ECE-Veg')
ds4.lai.weighted(weights).mean(['lon', 'lat']).plot(label = 'ACCESS')
plt.plot(ds4.time, ds3.lai.weighted(weights).mean(['lon', 'lat']).values, label = 'CESM2')
plt.plot(ds4.time, ds1m.isel(time = slice(12, 192)).weighted(weights).mean(['lon', 'lat']).values, linewidth = 3)

plt.legend()
```

```python
ds2.isel(time = slice(0, 12)).lai.weighted(weights).mean(['lon', 'lat']).plot(label = 'ECE-Veg')
ds4.isel(time = slice(0, 12)).lai.weighted(weights).mean(['lon', 'lat']).plot(label = 'ACCESS')
plt.plot(ds4.time[:12], ds3.isel(time = slice(0, 12)).lai.weighted(weights).mean(['lon', 'lat']).values, label = 'CESM2')
plt.plot(ds4.time[:12], ds1m.isel(time = slice(12, 24)).weighted(weights).mean(['lon', 'lat']).values, linewidth = 3, label = 'ESA satellite')

plt.legend()
```

```python
ds2.lai.median(['lon', 'lat']).plot(label = 'ECE-Veg')
ds4.lai.median(['lon', 'lat']).plot(label = 'ACCESS')
plt.plot(ds4.time, ds3.lai.median(['lon', 'lat']).values, label = 'CESM2')

#(ds1.lai_lv + ds1.lai_hv).weighted(weights).mean(['lon', 'lat']).plot(label = 'ERA5 Land sum')
ds1.lai_lv.median(['lon', 'lat']).plot(linewidth = 3, label = 'low')
ds1.lai_hv.median(['lon', 'lat']).plot(linewidth = 3, label = 'high')

plt.legend()
```

```python
ds2.lai.max(['lon', 'lat']).plot(label = 'ECE-Veg')
ds4.lai.max(['lon', 'lat']).plot(label = 'ACCESS')
plt.plot(ds4.time, ds3.lai.max(['lon', 'lat']).values, label = 'CESM2')

#(ds1.lai_lv + ds1.lai_hv).weighted(weights).mean(['lon', 'lat']).plot(label = 'ERA5 Land sum')
ds1.lai_lv.max(['lon', 'lat']).plot(linewidth = 3, label = 'low')
ds1.lai_hv.max(['lon', 'lat']).plot(linewidth = 3, label = 'high')

plt.legend()
```

```python
ds2.lai.min(['lon', 'lat']).plot(label = 'ECE-Veg')
ds4.lai.min(['lon', 'lat']).plot(label = 'ACCESS')
plt.plot(ds4.time, ds3.lai.min(['lon', 'lat']).values, label = 'CESM2')

#(ds1.lai_lv + ds1.lai_hv).weighted(weights).mean(['lon', 'lat']).plot(label = 'ERA5 Land sum')
ds1.lai_lv.min(['lon', 'lat']).plot(linewidth = 3, label = 'low')
ds1.lai_hv.min(['lon', 'lat']).plot(linewidth = 3, label = 'high')

plt.legend()
```
