```python
ds1 = xr.open_dataset('ESA_satellite/LAI_1999-2020_V2.0.1.nc_LAI_reg_arctic.nc')
ds2 = xr.open_dataset('ESA_satellite/NDVI_1999-2020_GLOBE_PROBAV_V3.0.1.nc_NDVI_reg_arctic.nc')
```

```python
ds1 = ds1.sel(lat = slice(60, 80))
ds2 = ds2.sel(lat = slice(60, 80))
```

```python
ds2['time'] =  ds2.time - np.timedelta64(1,'D')
```

```python
da1 = ds1.sel(time = (ds1.time.dt.month.isin(range(2, 8))) & (ds1.time.dt.year.isin(range(1999, 2015)))).LAI.bfill(dim = 'time')#.interpolate_na(dim = 'time', method = 'linear')#.dropna(dim = 'time')
da2 = ds2.sel(time = (ds2.time.dt.month.isin(range(2, 8))) & (ds2.time.dt.year.isin(range(1999, 2015)))).NDVI.bfill(dim = 'time')#.interpolate_na(dim = 'time', method = 'linear')#.dropna(dim = 'time')
```

```python
for yr in range(2000, 2015):
    maxim = da2.sel(time = da2.time.dt.year == yr).max(dim = 'time')
    minim = da2.sel(time = da2.time.dt.year == yr).min(dim = 'time')
    value = da2.sel(time = da2.time.dt.year == yr)

    exec(f"ndvi_perc_{yr} = 100*((value - minim)/(maxim - minim))")
```

```python
def nomi_poly_pixel(x, y, poly = PolynomialFeatures(degree = 3, include_bias=False), poly_reg_model = LinearRegression()):
    poly_reg_model = poly_reg_model
    poly = poly
    poly_features = poly.fit_transform(x.reshape(-1, 1))
#    df_reg = pd.DataFrame(poly_features)
#    df_reg['y'] = y
#    df_reg = df_reg.sort_values(by = 0)
    poly_reg_model.fit(poly_features, y)
    
    y_predicted = poly_reg_model.predict(poly_features)
    return y_predicted, poly_reg_model
```

```python
def nomi_poly2(x, y, deg):
    poly_reg_model = LinearRegression()
    poly = PolynomialFeatures(degree = deg, include_bias=False)
    poly_features = poly.fit_transform(x.reshape(-1, 1))
    df_reg = pd.DataFrame(poly_features)
    df_reg['y'] = y
    df_reg = df_reg.sort_values(by = 0)
    poly_reg_model.fit(df_reg.iloc[:, :deg], df_reg['y'])
    y_predicted = poly_reg_model.predict(df_reg.iloc[:, :deg])
    return y_predicted, poly_reg_model
```

```python
def nomi_rmse(y_pred, y):
    return np.sqrt(np.nanmean((y_pred - y)**2))
```

# Years

```python
def nomi_poly_years(x, y, deg):
    poly_reg_model = LinearRegression()
    poly = PolynomialFeatures(degree = deg, include_bias=False)
    poly_features = poly.fit_transform(x.reshape(-1, 1))
    df_reg = pd.DataFrame(poly_features)
    df_reg['y'] = y
    df_reg = df_reg.sort_values(by = 0)
    poly_reg_model.fit(df_reg.iloc[:, :deg], df_reg['y'])
    y_predicted = poly_reg_model.predict(df_reg.iloc[:, :deg])
    return y_predicted, df_reg, poly_reg_model
```

```python
def nomi_poly_predict(x, y, deg, model):
    poly_reg_model = model
    poly = PolynomialFeatures(degree = deg, include_bias=False)
    poly_features = poly.fit_transform(x.reshape(-1, 1))
    df_reg = pd.DataFrame(poly_features)
    df_reg['y'] = y
    df_reg = df_reg.sort_values(by = 0)
    #poly_reg_model.fit(df_reg.iloc[:, :deg], df_reg['y'])
    y_predicted = model.predict(df_reg.iloc[:, :deg])
    return y_predicted
```

```python
def nomi_poly_simple(x, a, b, c, d):
    y_predicted = a*x**3 + b*x**2 + c*x + d
    return y_predicted
```

```python
%%time
# Create the training dataset, containing 10 of the 15 years, and train the algorithm on it
year_train = np.random.choice(range(2000, 2014), 10, replace = False)
exec(f"x_train_da = da1.sel(time = da1.time.dt.year.isin(year_train)).values")
exec(f"y_train_da = da2_perc.sel(time = da2_perc.time.dt.year.isin(year_train)).values")

exec(f"ypred = np.zeros((x_train_da.shape[1], x_train_da.shape[2], x_train_da.shape[0]))")
exec(f"coef = np.zeros((x_train_da.shape[1], x_train_da.shape[2], 3))")
exec(f"interc = np.zeros((x_train_da.shape[1], x_train_da.shape[2]))")
#exec(f"rmse_{y} = np.zeros((len(da1_{y}.lat.values), len(da1_{y}.lon.values)))")
#exec(f"corr_{y} = np.zeros((len(da1_{y}.lat.values), len(da1_{y}.lon.values)))")

for j in range(x_train_da.shape[1]):
        for i in range(x_train_da.shape[2]):
            if eval(f"(np.sum(np.isnan(x_train_da[:, j, i])) < 18) & (np.sum(np.isnan(x_train_da[:, j, i])) == np.sum(np.isnan(y_train_da[:, j, i])))"):        
                exec(f"x_train = x_train_da[:, j, i]")
                #exec(f"print({y, i, j}, x_test)")
                exec(f"y_train = y_train_da[:, j, i]")
                y_predicted, df_train, poly_reg_model = nomi_poly_years(x_train, y_train, 3)
                exec(f"ypred[{j}, {i}, :] = y_predicted")
                exec(f"coef[j, i, :] = poly_reg_model.coef_")
                exec(f"interc[j, i] = poly_reg_model.intercept_")
                #exec(f"rmse_{y}[j, i] = nomi_rmse(y_predicted, y_test)")
                #exec(f"corr_{y}[j, i] = np.corrcoef(y_predicted, y_test)[0][1]")
            else:
                #print("No match")
                exec(f"ypred[j, i, :] = np.nan")
                exec(f"coef[j, i, :] = np.nan")
                exec(f"interc[j, i] = np.nan")
                #exec(f"rmse_{y}[j, i] = np.nan")
                #exec(f"corr_{y}[j, i] = np.nan")
```

```python
%%time
#Create the testing dataset, containing 5 of the 15 years, and test the algorithm on it 5 times (one for each year)
year_test = list(range(2000, 2014))
year_test = [ele for ele in year_test if ele not in year_train]
for yr in range(4):
    exec(f"x_test_{yr+1} = da1.sel(time = da1.time.dt.year.isin(year_test[{yr}])).values")
    exec(f"y_test_{yr+1} = da2_perc.sel(time = da2_perc.time.dt.year.isin(year_test[{yr}])).values")
    exec(f"y_pred_{yr+1} = np.zeros((x_test_{yr+1}.shape[1], x_test_{yr+1}.shape[2], x_test_{yr+1}.shape[0]))")
    for j in eval(f"range(x_test_{yr+1}.shape[1])"):
            for i in eval(f"range(x_test_{yr+1}.shape[2])"):
                if eval(f"(np.sum(np.isnan(x_test_{yr+1}[:, j, i])) < 18) & (np.sum(np.isnan(x_test_{yr+1}[:, j, i])) == np.sum(np.isnan(y_test_{yr+1}[:, j, i])))"):        
                    exec(f"x_test = x_test_{yr+1}[:, j, i]")
                    #exec(f"print({y, i, j}, x_test)")
                    exec(f"y_test = y_test_{yr+1}[:, j, i]")
                    y_predicted = nomi_poly_simple(x_test, coef[j, i, 2], coef[j, i, 1], coef[j, i, 0], interc[j, i])
                    exec(f"y_pred_{yr+1}[{j}, {i}, :] = y_predicted")
                    #exec(f"rmse_{y}[j, i] = nomi_rmse(y_predicted, y_test)")
                    #exec(f"corr_{y}[j, i] = np.corrcoef(y_predicted, y_test)[0][1]")
                else:
                    #print("No match")
                    exec(f"y_pred_{yr+1}[{j}, {i}, :] = np.nan")
                    #exec(f"rmse_{y}[j, i] = np.nan")
                    #exec(f"corr_{y}[j, i] = np.nan")
```
